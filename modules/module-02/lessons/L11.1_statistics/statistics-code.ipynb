## Variance and Standard Deviation
Variance is the average squared distance from the mean. Standard deviation is its square root — it
brings the unit back to the original scale of your data, making it interpretable.

```python
import numpy as np
data=np.array([10, 12, 14, 16, 18, 20])

mean=np.mean(data)
```

```python
print(mean)

# Manual variance calculation - from first principles
deviation=data-mean
deviation
```

```python
# numpy nuild in ddof=0 for population , ddof=1 for sample
print(f"Population variance {np.var(data, ddof=0):.4f}")
print(f"Sample variance:  {np.var(data, ddof=1):.4f}")
print(f'Std deviation (pop): {np.std(data, ddof=0):.4f}')
print(f'Std deviation (smp): {np.std(data, ddof=1):.4f}')
```

```python
sq_devs=deviation ** 2
sq_devs
```

```python
variance_pop=np.mean(sq_devs)
```

```python
print(f"Variance is :  {(variance_pop):.2f}")
```

```python
std_pop=np.sqrt(variance_pop)
```

```python
std_pop
```

## 1.5 Covariance and Correlation

### � Correlation
Measures the strength and direction of the linear relationship between two variables.
Ranges from -1 (perfect negative) to +1 (perfect positive). Zero means no linear
relationship — but not necessarily no relationship at all.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(7)
n=200
```

```python
print(np.random.seed(7))
```

```python
# create correlated variables

x=np.random.normal(0, 1, n)
print(x)
```

```python
y_positive=0.8 * x + 0.2*np.random.normal(0, 1,n)
print(y_positive)
```
