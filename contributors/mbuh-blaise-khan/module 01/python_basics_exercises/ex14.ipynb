```python
# Define function with default parameters
def report_metrics(model_name, accuracy, loss, verbose=True, round_to=4):
    """
    Print evaluation report for a model.
    
    Parameters:
    - model_name: Name of the model
    - accuracy: Model accuracy (float)
    - loss: Model loss (float)
    - verbose: If True, print all metrics; if False, print only name and accuracy
    - round_to: Number of decimal places to round to (default 4)
    """
    # Round the metrics
    accuracy = round(accuracy, round_to)
    loss = round(loss, round_to)
    
    if verbose:
        print("=" * 40)
        print(f"MODEL EVALUATION REPORT: {model_name}")
        print("=" * 40)
        print(f"Accuracy: {accuracy}%")
        print(f"Loss: {loss}")
        print(f"Rounding precision: {round_to} decimals")
        print("=" * 40)
    else:
        print(f"{model_name}: {accuracy}% (loss: {loss})")
    
    return None  # Function just prints, doesn't return anything

# Call 1: All arguments specified
print("1. All arguments specified:")
report_metrics("ResNet50", 94.56789, 0.23456, True, 3)

# Call 2: Rely on defaults (verbose=True, round_to=4)
print("\n2. Relying on defaults:")
report_metrics("EfficientNet", 92.34567, 0.34567)

# Call 3: Only change verbose to False
print("\n3. Only verbose=False:")
report_metrics("MobileNet", 88.99999, 0.45678, verbose=False)

# Call 4: Only change round_to to 2 (using keyword argument)
print("\n4. Different round_to value (2 decimals):")
report_metrics("ViT", 96.123456, 0.123456, round_to=2)

# Bonus: Show mixing positional and keyword arguments
print("\n5. Mixed arguments (verbose=False, round_to=1):")
report_metrics("Bert", 97.654321, 0.098765, False, 1)
```
