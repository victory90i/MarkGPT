```python
model_config = {
    'architecture': 'ResNet50',
    'learning_rate': 0.001,
    'batch_size': 32,
    'epochs': 100,
    'optimizer': 'Adam',
    'dropout_rate': 0.5
}

print("=== Original Model Configuration ===")
# (1) Print every key-value pair using .items()
for key, value in model_config.items():
    print(f"{key:20}: {value}")
    
# (2) Update learning_rate to 0.0001
model_config['learning_rate'] = 0.0001

# (3) Add new key early_stopping = True
model_config['early_stopping'] = True

# (4) Safely retrieve 'regularisation' with default 'none'
regularisation = model_config.get('regularisation', 'none')

print("\n=== Updated Model Configuration ===")
for key, value in model_config.items():
    print(f"{key:20}: {value}")

print(f"\nRegularisation value (not in config): {regularisation}")

# Demonstrate .get() safety
print("\n=== Demonstrating .get() Safety ===")
# This would crash with KeyError: print(model_config['regularisation'])
# But .get() returns default safely:
missing_key = model_config.get('does_not_exist', 'default_value')
print(f"Missing key returns: '{missing_key}'")
```
