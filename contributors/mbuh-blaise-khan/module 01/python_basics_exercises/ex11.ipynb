```python
# Cell 1: Celsius to Kelvin function
def celsius_to_kelvin(c):
    """Convert Celsius to Kelvin."""
    return c + 273.15

# Cell 2: Min-Max Normalisation function
def normalise(value, min_val, max_val):
    """Apply min-max normalisation to scale value between 0 and 1."""
    return (value - min_val) / (max_val - min_val)

# Cell 3: Apply both functions to temperature readings
temp_celsius = [10, 25, 37, 100, -5]

# Convert all to Kelvin first
temp_kelvin = [celsius_to_kelvin(c) for c in temp_celsius]
print("Temperatures in Kelvin:", temp_kelvin)

# Find min and max for normalisation
min_kelvin = min(temp_kelvin)
max_kelvin = max(temp_kelvin)
print(f"Min Kelvin: {min_kelvin}, Max Kelvin: {max_kelvin}\n")

# Normalise the Kelvin values
normalised_temps = [normalise(k, min_kelvin, max_kelvin) for k in temp_kelvin]

# Display results side by side
print("Original (°C) → Kelvin → Normalised")
for c, k, n in zip(temp_celsius, temp_kelvin, normalised_temps):
    print(f"{c:6} → {k:8.2f} → {n:.4f}")
```
