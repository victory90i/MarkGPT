```python
def get_batches(data, batch_size):
    # Loop through indices: 0, batch_size, 2*batch_size, etc.
    for start_index in range(0, len(data), batch_size):
        # Slice from start_index to start_index + batch_size
        batch = data[start_index:start_index + batch_size]
        yield batch  # Return this batch and PAUSE

# Test with dataset of 10 numbers (1 to 10)
dataset = list(range(1, 11))

# Test batch size = 3
print("=== Batch Size 3 ===")
for batch_num, batch in enumerate(get_batches(dataset, 3), start=1):
    print(f"Batch {batch_num}: {batch}")

# Test batch size = 4
print("\n=== Batch Size 4 ===")
for batch_num, batch in enumerate(get_batches(dataset, 4), start=1):
    print(f"Batch {batch_num}: {batch}")
```
