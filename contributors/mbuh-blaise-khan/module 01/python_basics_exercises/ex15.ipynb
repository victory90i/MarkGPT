```python
# Define global variables
best_val_accuracy = 0.0
checkpoint_log = []

def update_checkpoint(epoch, val_acc, model_name):
    """
    Update the best validation accuracy and log checkpoints.
    
    Args:
        epoch: Current epoch number
        val_acc: Validation accuracy for this epoch
        model_name: Name of the model being trained
    """
    global best_val_accuracy  # CRITICAL: tells Python to use global variable
    
    # Check if current accuracy is better than best
    if val_acc > best_val_accuracy:
        # Update global best accuracy
        best_val_accuracy = val_acc
        
        # Log this checkpoint
        checkpoint_entry = {
            'epoch': epoch,
            'accuracy': val_acc,
            'model': model_name,
            'is_best': True
        }
        checkpoint_log.append(checkpoint_entry)
        print(f"✓ NEW BEST! Epoch {epoch}: {val_acc:.4f} (previous best: {best_val_accuracy - val_acc + val_acc:.4f})")
    else:
        # Log without updating best
        checkpoint_entry = {
            'epoch': epoch,
            'accuracy': val_acc,
            'model': model_name,
            'is_best': False
        }
        checkpoint_log.append(checkpoint_entry)
        print(f"  Epoch {epoch}: {val_acc:.4f} (no improvement)")

# Simulate training for 5 epochs with a list of validation accuracies
print("=== Starting Training Simulation ===\n")
validation_accuracies = [0.85, 0.88, 0.92, 0.91, 0.93]

for epoch, acc in enumerate(validation_accuracies, start=1):
    update_checkpoint(epoch, acc, "MyNeuralNetwork")

# Print final results
print("\n=== Final Training Summary ===")
print(f"Best validation accuracy achieved: {best_val_accuracy:.4f}")
print(f"Total checkpoints logged: {len(checkpoint_log)}")

print("\n=== Checkpoint Log ===")
for entry in checkpoint_log:
    status = "🏆 BEST" if entry['is_best'] else "  Saved"
    print(f"  Epoch {entry['epoch']}: {entry['accuracy']:.4f} - {status}")

# Demonstrate what happens WITHOUT the global keyword
print("\n=== DEMONSTRATION: Why 'global' is needed ===")

bad_accuracy = 0.0  # Global variable

def bad_update(accuracy):
    bad_accuracy = accuracy  # This creates a LOCAL variable!
    print(f"Inside function: bad_accuracy = {bad_accuracy}")

print(f"Before function: bad_accuracy = {bad_accuracy}")
bad_update(0.95)
print(f"After function: bad_accuracy = {bad_accuracy}")  # Still 0.0!
print("❌ Without 'global', the global variable NEVER changes!")
```
