```python
def validate_sample(age, income, credit_score):
    # Check for negative values first (any one negative fails)
    if age < 0 or income < 0 or credit_score < 0:
        return "Invalid: negative values not allowed"
    
    # Check age range
    if age < 18 or age > 100:
        return "Invalid: age out of range"
    
    # Check unusually high income
    if income > 1_000_000:   # underscore for readability (1 million)
        return "Flag: unusually high income — review"
    
    # All checks passed
    return "Valid sample"

# Test cases (run each in a separate cell or one after the other)
print(validate_sample(25, 50000, 720))      # Valid sample
print(validate_sample(-5, 50000, 720))      # Invalid: negative...
print(validate_sample(15, 50000, 720))      # Invalid: age out of range
print(validate_sample(30, 2_000_000, 800))  # Flag: unusually high income
```
