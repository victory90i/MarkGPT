Q1. What is the key difference between Matplotlib and Seaborn? When would you choose
one over the other?

Matplotlib provides basic plotting capabilities; Seaborn builds on it with prettier, more complex visualizations and statistical features. Use Matplotlib for customization and control; choose Seaborn for quick, attractive statistical plots

Q5. What does sns.set_theme(style='whitegrid') do? Try changing the style to 'darkgrid' or
'ticks' and observe the difference

`sns.set_theme(style='whitegrid')` sets the visual style of Seaborn plots to have a clean white background with gridlines, making it easier to read data points.

**Changing styles:**

- `'darkgrid'`: Darker background with gridlines, providing a more subdued look, good for presentations.
- `'ticks'`: White background with ticks on axes, minimal gridlines, giving a cleaner, simpler appearance.

**Observation:**  
Switching styles changes the plot's background, gridline appearance, and overall aesthetic, which can help highlight different aspects of your data depending on your visualization needs.
