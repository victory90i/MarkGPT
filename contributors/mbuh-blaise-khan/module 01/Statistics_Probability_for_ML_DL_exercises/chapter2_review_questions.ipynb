Q1. Disease Testing and Bayes' Theorem
A disease affects 0.5% of the population. A test correctly identifies 98% of cases
(sensitivity) and correctly clears 95% of healthy people (specificity). Using Bayes' theorem,
calculate the probability that a person actually has the disease given a positive test result.
What does your answer tell you about testing for rare conditions?

Given:  

Disease prevalence (P(D)) = 0.005 (0.5%)  
Sensitivity = 98% = 0.98  
Specificity = 95% = 0.95


Calculate: Probability of having the disease given a positive test (P(D|Pos)) using Bayes' theorem:


\(P(D|Pos) = \frac{P(Pos|D) \times P(D)}{P(Pos|D) \times P(D) + P(Pos|~D) \times P(~D)}\)
Where:  

\(P(Pos|D) = 0.98\)  
\(P(Pos|~D) = 1 - \text{Specificity} = 0.05\)  
\(P(~D) = 1 - 0.005 = 0.995\)

Plugging in:\(P(D|Pos) = \frac{0.98 \times 0.005}{0.98 \times 0.005 + 0.05 \times 0.995} \approx 0.047\)
Interpretation: About 4.7% chance someone who tests positive actually has the disease.Insight: For rare conditions, even good tests produce many false positives, so positive results should be interpreted cautiously.

Q2. What is the difference between P(A and B) and P(A | B)? Give an example where
confusing these two would lead to a wrong conclusion in an ML context.

P(A and B) vs. P(A | B)  

P(A and B): Probability both A and B occur.  
P(A | B): Probability A occurs given B has occurred.

Example:In ML, confusing these could lead to errors like:  

Mistaking the probability of both features occurring (P(A and B)) for the probability of A given B (P(A|B)).  
This could cause incorrect assumptions about feature dependencies or class probabilities

Q4 What is the Law of Large Numbers? Why does it give us confidence that simulated
probability estimates converge to true probabilities with enough samples?
States that as the number of trials increases, the average of results approaches the expected value.  
Why: It ensures that with enough samples, our simulated estimates of probability will be close to the true probabilities, providing confidence in simulation results.

Q5. Explain the base rate fallacy in your own words. Why is accuracy a misleading metric
for highly imbalanced classification problems?Q5. Base Rate Fallacy & Accuracy  

Base Rate Fallacy: Ignoring the actual prevalence (base rate) of a condition when interpreting test results, leading to incorrect conclusions.  
Why accuracy is misleading: In imbalanced data (e.g., rare diseases), a model can have high accuracy by always predicting the majority class, but it fails to identify the minority class effectively.
