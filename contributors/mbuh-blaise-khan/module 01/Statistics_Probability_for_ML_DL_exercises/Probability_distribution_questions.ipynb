Q1. A neural network outputs sigmoid activations for binary classification. If the model
predicts P(cat) = 0.82 for an image, and the true label is 1 (cat), compute the binary cross-
entropy loss for this single sample: -[y*log(p) + (1-y)*log(1-p)]. Then explain the connection
to the Bernoulli distribution

```python

Neural network sigmoid activation and Bernoulli connection

Given: The model predicts \(P(\text{cat}) = 0.82\); true label \(y=1\).

Binary cross-entropy loss:\(L = -[ y \log p + (1 - y) \log (1 - p) ]\)

Calculation:\(L = -[1 \times \log 0.82 + 0 \times \log (1 - 0.82)] = - \log 0.82 \approx -(-0.198) = 0.198\)
# Connection to Bernoulli distribution:In a Bernoulli distribution, the probability of success (1) is \(p\); the loss function used in training (cross-entropy) is derived from the negative log-likelihood of observing the data under this Bernoulli model.  

# Why it makes sense:Maximizing likelihood (or minimizing negative log-likelihood) for Bernoulli trials (binary outcomes) leads to the cross-entropy loss. The model's predicted probability is the Bernoulli parameter, and the loss quantifies how well the predicted probability matches the actual outcome.
```

Q2. Using scipy.stats.norm, find the value x such that P(X ≤ x) = 0.975 for a standard
normal distribution. This value (≈ 1.96) is extremely important in hypothesis testing —
explain why.

```python
import scipy.stats as stats
x = stats.norm.ppf(0.975)
print(x)
Result: \(x \approx 1.96\)

# Why important:In hypothesis testing, this value represents the critical value for a 95% confidence level. If a test statistic exceeds 1.96 (or is less than -1.96), the result is considered statistically significant at the 5% level, indicating strong evidence against the null hypothesis.
```

Q3. Simulate the Central Limit Theorem using a uniform distribution U(0, 1) instead of
exponential. Draw 10,000 sample means for samples of size n=1, n=5, and n=30. Plot
histograms. At what sample size does the distribution of means look approximately normal?

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Set parameters
num_samples = 10000
sample_sizes = [1, 5, 30]

# 2. Create subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=False)

# 3. Generate data, compute means, and plot histograms
for ax, n in zip(axes, sample_sizes):
    # Generate a 2D array of uniform samples (num_samples rows, n columns)
    samples = np.random.uniform(low=0.0, high=1.0, size=(num_samples, n))
    
    # Compute the mean of each sample (across the rows)
    sample_means = np.mean(samples, axis=1)
    
    # Plot the histogram of the means
    ax.hist(sample_means, bins=50, density=True, color='steelblue', edgecolor='white', alpha=0.8)
    
    # Formatting
    ax.set_title(f'Sample Size n = {n}')
    ax.set_xlabel('Sample Mean')
    ax.set_ylabel('Density')
    ax.grid(axis='y', alpha=0.3)

plt.suptitle('Central Limit Theorem: Convergence to Normal Distribution', fontsize=14, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()


# Approach:Generate 10,000 samples of sizes \(n=1, 5, 30\). For each, compute the mean of each sample, then plot histograms of these means.

# Expected result:  

# For \(n=1\), the distribution of sample means is just the original uniform distribution, not normal.

# For larger \(n\), especially \(n=30\), the distribution of the sample means will look approximately Gaussian due to CLT.

# Answer:The distribution of means begins to look approximately normal around \(n=30\).
```

Q4. Xavier initialisation uses weights drawn from Uniform(-sqrt(6/(fan_in+fan_out)),
sqrt(6/(fan_in+fan_out))). Using NumPy, compute the theoretical variance of Xavier-
initialised weights for a layer with fan_in=128 and fan_out=64. What does this variance
ensure about signal propagation?

```python
Xavier initialization variance calculation

Formula:\(\text{Variance} = \frac{(b - a)^2}{12}\)

Given:a = -\sqrt{\frac{6}{\text{fan_in} + \text{fan_out}}} \\
b = +\sqrt{\frac{6}{\text{fan_in} + \text{fan_out}}}

Compute for:\text{fan_in} = 128, \quad \text{fan_out} = 64
Calculation:\(limit = \sqrt{\frac{6}{128 + 64}} = \sqrt{\frac{6}{192}} \approx \sqrt{0.03125} \approx 0.177\)Variance of weights uniformly distributed between \(-limit\) and \(+limit\):\(\text{Var} = \frac{(2 \times limit)^2}{12} = \frac{(2 \times 0.177)^2}{12} = \frac{(0.354)^2}{12} \approx \frac{0.125}{12} \approx 0.0104\)

Implication:This variance ensures signals neither vanish nor explode as they propagate through the network, facilitating stable training.
```

Q5. What is the difference between a PDF and a PMF? Give one example of a distribution
that uses a PDF and one that uses a PMF from this chapter.

```python
Difference between PDF and PMF

PDF (Probability Density Function):  

Applies to continuous distributions (e.g., normal, uniform).  

Gives the density at each point; the probability of exactly any single point is zero, but the area under the curve over an interval gives the probability.

PMF (Probability Mass Function):  

Applies to discrete distributions (e.g., Bernoulli, Binomial, Poisson).  

Gives the probability that a random variable equals a specific value.

Examples from chapter:  

PDF: Normal distribution (e.g., stats.norm.pdf(x))  
PMF: Poisson distribution (e.g., stats.poisson.pmf(k))
```
