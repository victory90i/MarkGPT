```python
# Define feature sets from two sources
features_source_a = {'age', 'income', 'credit_score', 'loan_amount', 'zip_code'}
features_source_b = {'age', 'income', 'employment_status', 'zip_code', 'region'}

print("=" * 60)
print("FEATURE SET ANALYSIS")
print("=" * 60)

# Original sets
print(f"\nSource A features ({len(features_source_a)}): {features_source_a}")
print(f"Source B features ({len(features_source_b)}): {features_source_b}")

# (1) All unique features across both sources - UNION (|)
all_features = features_source_a | features_source_b
print(f"\n1. ALL UNIQUE FEATURES (Union - |): {len(all_features)} features")
print(f"   {all_features}")

# (2) Features that appear in both - INTERSECTION (&)
common_features = features_source_a & features_source_b
print(f"\n2. FEATURES IN BOTH (Intersection - &): {len(common_features)} features")
print(f"   {common_features}")
print(f"   ✓ These are safe to merge without duplication")

# (3) Features exclusive to each source - DIFFERENCE (-)
exclusive_to_a = features_source_a - features_source_b
exclusive_to_b = features_source_b - features_source_a
print(f"\n3. FEATURES EXCLUSIVE TO EACH SOURCE (Difference -):")
print(f"   Only in Source A ({len(exclusive_to_a)}): {exclusive_to_a}")
print(f"   Only in Source B ({len(exclusive_to_b)}): {exclusive_to_b}")

# (4) Features that are in one source but not both - SYMMETRIC DIFFERENCE (^)
one_source_only = features_source_a ^ features_source_b
print(f"\n4. FEATURES IN ONE SOURCE BUT NOT BOTH (Symmetric Difference - ^):")
print(f"   {len(one_source_only)} features: {one_source_only}")
print(f"   (This is the same as exclusive_to_a ∪ exclusive_to_b)")

```
