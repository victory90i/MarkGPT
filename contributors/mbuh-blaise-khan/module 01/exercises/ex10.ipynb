```python
def tokenize(text):
    # Step 1: Convert to lowercase
    text = text.lower()
    
    # Step 2: Remove punctuation (replace with spaces)
    # Common punctuation to remove: . , ! ? ; : ( ) " '
    punctuation = ".,!?;:()\"'"
    for punc in punctuation:
        text = text.replace(punc, " ")
    
    # Step 3: Split into words
    words = text.split()
    
    # Step 4: Define stop words (use a set for fast lookup)
    stop_words = {'a', 'an', 'the', 'is', 'in', 'of', 'and', 'to'}
    
    # Step 5: Filter out stop words
    tokens = []
    for word in words:
        if word not in stop_words:
            tokens.append(word)
    
    return tokens

# Test the function
sentence = 'The quick brown fox jumps over the lazy dog and the cat.'
tokens = tokenize(sentence)

print("Tokens:", tokens)

# Get vocabulary (unique tokens)
vocab = sorted(set(tokens))
print("Vocabulary:", vocab)

# Get frequency
freq = {}
for token in tokens:
    freq[token] = freq.get(token, 0) + 1
print("Frequency:", freq)
```
