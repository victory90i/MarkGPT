```python
val_losses = [0.45, 0.38, 0.34, 0.34, 0.34, 0.34, 0.33]
patience = 3
epoch = 0
count = 0
best_loss = val_losses[0]

while epoch < len(val_losses):
    current_loss = val_losses[epoch]
    if current_loss <= best_loss:
        best_loss = current_loss
        print(f"Epoch {epoch}: current_loss improved to {current_loss}")
    else:
        count += 1
        print(f"Epoch {epoch}: no improvement (counter = {counter})")
    if count == patience:
        print (f"Early stopping triggered at epoch {epoch} count: {count}")
        break
    else: 
        print("Finished all epochs without early stopping")

    epoch += 1
```
