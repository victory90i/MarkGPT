```python
# Create the sample row and column names
sample = [28, 'engineer', 85000, True, 720]
column_names = ['age', 'occupation', 'income', 'employed', 'credit_score']

print("=== Original Dataset Row ===")
# (1) Print each field with its column name using zip()
for name, value in zip(column_names, sample):
    print(f"{name:15}: {value}")

# (2) Append a new field loan_approved = False
sample.append(False)
print(f"\nAfter appending 'loan_approved': {sample}")

# (3) Remove the 'employed' field by value (True)
# Note: We need to find the index of True in the sample list
# But careful: True appears at index 3 (the 'employed' field)
sample.remove(True)
print(f"After removing 'employed' field: {sample}")

# Also update column_names to reflect changes
column_names.append('loan_approved')
# Remove 'employed' from column_names by its index (position 3)
column_names.remove('employed')

print("\n=== Final Dataset Row with Updated Columns ===")
for name, value in zip(column_names, sample):
    print(f"{name:15}: {value}")
```
