```python
raw_scores = [15, -3, 200, 42, 0, 99, -7, 55, 188, 30]

# 1. Clip values to [0, 100]
clipped_scores = [max(0, min(x, 100)) for x in raw_scores]
print("Clipped:", clipped_scores)

# 2. Normalise each clipped value to [0, 1] by dividing by 100
normalised = [x / 100 for x in clipped_scores]
print("Normalised:", normalised)

# 3. Create labels: 'high' if normalised >= 0.6 else 'low'
labels = ['high' if val >= 0.6 else 'low' for val in normalised]
print("Labels :", labels)
```
