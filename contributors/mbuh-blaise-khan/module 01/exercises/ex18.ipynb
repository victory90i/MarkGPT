```python
# Initialize the experiment tracker (empty dictionary)
experiment_tracker = {}

def log_run(run_id, params, metrics):
    """
    Store a run's hyperparameters and metrics.
    
    Args:
        run_id: Unique identifier for the run (string)
        params: Dictionary of hyperparameters
        metrics: Dictionary of evaluation metrics
    """
    experiment_tracker[run_id] = {
        'params': params,
        'metrics': metrics
    }
    print(f"✓ Logged run '{run_id}'")

def get_best_run(metric_name):
    """
    Find the run with the highest value for a given metric.
    
    Args:
        metric_name: Name of the metric to optimize (e.g., 'accuracy')
    
    Returns:
        Tuple of (run_id, metric_value)
    """
    if not experiment_tracker:
        return None, None
    
    best_run_id = None
    best_value = -float('inf')
    
    for run_id, data in experiment_tracker.items():
        if metric_name in data['metrics']:
            current_value = data['metrics'][metric_name]
            if current_value > best_value:
                best_value = current_value
                best_run_id = run_id
    
    return best_run_id, best_value

def summarise():
    """
    Print a formatted table of all logged runs.
    """
    if not experiment_tracker:
        print("No runs logged yet.")
        return
    
    print("\n" + "=" * 90)
    print("EXPERIMENT TRACKER SUMMARY")
    print("=" * 90)
    
    # Print header
    print(f"{'Run ID':<15} {'Learning Rate':<12} {'Batch Size':<10} {'Epochs':<8} {'Accuracy':<10} {'Loss':<10}")
    print("-" * 90)
    
    # Print each run
    for run_id, data in experiment_tracker.items():
        params = data['params']
        metrics = data['metrics']
        
        lr = params.get('learning_rate', 'N/A')
        batch = params.get('batch_size', 'N/A')
        epochs = params.get('epochs', 'N/A')
        acc = metrics.get('accuracy', 'N/A')
        loss = metrics.get('loss', 'N/A')
        
        # Format numbers nicely
        if isinstance(lr, float):
            lr = f"{lr:.5f}"
        if isinstance(acc, float):
            acc = f"{acc:.4f}"
        if isinstance(loss, float):
            loss = f"{loss:.4f}"
        
        print(f"{run_id:<15} {lr:<12} {batch:<10} {epochs:<8} {acc:<10} {loss:<10}")
    
    print("=" * 90)

# Log at least four fake runs
print("📝 LOGGING EXPERIMENT RUNS")
print("-" * 40)

log_run('run_001', 
        {'learning_rate': 0.01, 'batch_size': 32, 'epochs': 50},
        {'accuracy': 0.85, 'loss': 0.45})

log_run('run_002', 
        {'learning_rate': 0.001, 'batch_size': 64, 'epochs': 100},
        {'accuracy': 0.92, 'loss': 0.23})

log_run('run_003', 
        {'learning_rate': 0.0001, 'batch_size': 128, 'epochs': 150},
        {'accuracy': 0.94, 'loss': 0.18})

log_run('run_004', 
        {'learning_rate': 0.0005, 'batch_size': 32, 'epochs': 200},
        {'accuracy': 0.91, 'loss': 0.27})

# Display summary table
summarise()

# Find and display the best run for accuracy
print("\n🏆 FINDING BEST RUN")
print("-" * 40)
best_run_id, best_accuracy = get_best_run('accuracy')
if best_run_id:
    print(f"Best accuracy: {best_accuracy:.4f} from run '{best_run_id}'")
    print(f"Configuration: {experiment_tracker[best_run_id]['params']}")
    print(f"Metrics: {experiment_tracker[best_run_id]['metrics']}")

# Find best run for loss (lower is better - custom implementation)
print("\n📉 FINDING BEST RUN FOR LOSS (lower is better)")
print("-" * 40)

def get_best_run_min(metric_name):
    """Find run with lowest value for a metric."""
    if not experiment_tracker:
        return None, None
    
    best_run_id = None
    best_value = float('inf')
    
    for run_id, data in experiment_tracker.items():
        if metric_name in data['metrics']:
            current_value = data['metrics'][metric_name]
            if current_value < best_value:
                best_value = current_value
                best_run_id = run_id
    
    return best_run_id, best_value

best_loss_run, best_loss = get_best_run_min('loss')
if best_loss_run:
    print(f"Best loss: {best_loss:.4f} from run '{best_loss_run}'")

# Demonstrate accessing nested data
print("\n" + "=" * 50)
print("NESTED DATA STRUCTURE VISUALIZATION")
print("=" * 50)
print("experiment_tracker = {")
for run_id, data in experiment_tracker.items():
    print(f"    '{run_id}': {{")
    print(f"        'params': {data['params']},")
    print(f"        'metrics': {data['metrics']}")
    print(f"    }},")
print("}")
```
