```python
# Cell 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

print("✅ Libraries imported successfully")
```

```python
# Cell 2: Visualise vector addition in 2D
def plot_vector_addition():
    """Demonstrate vector addition with arrows"""
    
    # Define two vectors
    v1 = np.array([3, 2])   # Vector 1: right 3, up 2
    v2 = np.array([1, 3])   # Vector 2: right 1, up 3
    v_sum = v1 + v2          # Vector sum: [4, 5]
    
    # Create figure
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Plot vectors from origin
    ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', 
              scale=1, color='blue', width=0.015, label='v1 = (3,2)')
    ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', 
               scale=1, color='green', width=0.015, label='v2 = (1,3)')
    
    # Plot sum vector
    ax.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', 
               scale=1, color='red', width=0.02, label=f'v1+v2 = ({v_sum[0]},{v_sum[1]})')
    
    # Show parallelogram law (dashed lines)
    ax.plot([v1[0], v_sum[0]], [v1[1], v_sum[1]], 'b--', alpha=0.5)
    ax.plot([v2[0], v_sum[0]], [v2[1], v_sum[1]], 'g--', alpha=0.5)
    
    # # Formatting
    ax.set_xlim(-1, 6)
    ax.set_ylim(-1, 7)
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.grid(True, alpha=0.4)
    ax.set_aspect('equal')
    ax.set_title('Vector Addition: Head-to-Tail Method', fontsize=14, fontweight='bold')
    ax.legend(loc='upper left')
    
    plt.tight_layout()
    plt.show()
    
    # Print numerical results
    print(f"\n📊 Numerical Results:")
    print(f"   v1 = {v1}")
    print(f"   v2 = {v2}")
    print(f"   v1 + v2 = {v_sum}")
    print(f"   |v1| = {np.linalg.norm(v1):.3f}")
    print(f"   |v2| = {np.linalg.norm(v2):.3f}")
    print(f"   |v_sum| = {np.linalg.norm(v_sum):.3f}")

# Run the visualisation
plot_vector_addition()
```

```python
# Cell 3: Demonstrate dot product and cosine similarity
def explain_dot_product():
    """Explain dot product geometrically"""
    
    # Define two vectors
    a = np.array([4, 3])    # Length 5 vector (3-4-5 triangle)
    b = np.array([5, 0])    # Horizontal vector
    
    # Calculate dot product
    dot_ab = np.dot(a, b)
    
    # Calculate magnitudes
    mag_a = np.linalg.norm(a)
    mag_b = np.linalg.norm(b)
    
    # Calculate cosine similarity
    cos_sim = dot_ab / (mag_a * mag_b)
    angle_rad = np.arccos(cos_sim)
    angle_deg = np.degrees(angle_rad)
    
    # Create visualisation
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Plot vectors
    ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', 
              scale=1, color='blue', width=0.02, label=f'a = {a}')
    ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', 
              scale=1, color='green', width=0.02, label=f'b = {b}')
    
    # Draw angle arc
    theta = np.linspace(0, angle_rad, 50)
    r = 1.5
    arc_x = r * np.cos(theta)
    arc_y = r * np.sin(theta)
    ax.plot(arc_x, arc_y, 'r--', linewidth=2, label=f'Angle: {angle_deg:.1f}°')
    
    # Formatting
    ax.set_xlim(-1, 6)
    ax.set_ylim(-1, 5)
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')
    ax.set_title('Dot Product: a·b = |a||b|cos(θ)', fontsize=14, fontweight='bold')
    ax.legend(loc='upper right')
    
    plt.tight_layout()
    plt.show()
    
    # Print explanation
    print("\n" + "="*60)
    print("📐 DOT PRODUCT EXPLANATION")
    print("="*60)
    print(f"a = {a}, |a| = {mag_a:.3f}")
    print(f"b = {b}, |b| = {mag_b:.3f}")
    print(f"\nDot product a·b = {a[0]}×{b[0]} + {a[1]}×{b[1]} = {dot_ab}")
    print(f"\nCosine similarity = (a·b) / (|a|×|b|) = {dot_ab} / ({mag_a:.3f}×{mag_b:.3f}) = {cos_sim:.4f}")
    print(f"\nAngle between vectors = arccos({cos_sim:.4f}) = {angle_deg:.1f}°")
    print("\n💡 Interpretation:")
    if cos_sim > 0.9:
        print("   → Vectors point in nearly the SAME direction (small angle)")
    elif cos_sim < -0.9:
        print("   → Vectors point in nearly OPPOSITE directions")
    elif abs(cos_sim) < 0.1:
        print("   → Vectors are nearly PERPENDICULAR (independent)")
    else:
        print("   → Vectors have moderate alignment")

explain_dot_product()
```

```python
# Cell 4: 3D vector operations
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_vectors():
    """Visualise vectors in 3D space"""
    
    # Define 3D vectors
    v1 = np.array([2, 1, 3])
    v2 = np.array([1, 3, 1])
    v_sum = v1 + v2
    
    # Create 3D figure
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot vectors from origin
    # Quiver in 3D: (x, y, z, u, v, w)
    ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='blue', 
              arrow_length_ratio=0.1, linewidth=2, label='v1')
    ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='green', 
              arrow_length_ratio=0.1, linewidth=2, label='v2')
    ax.quiver(0, 0, 0, v_sum[0], v_sum[1], v_sum[2], color='red', 
              arrow_length_ratio=0.1, linewidth=2.5, label='v1+v2')
    
    # Draw dashed lines for parallelogram
    ax.plot([v1[0], v_sum[0]], [v1[1], v_sum[1]], [v1[2], v_sum[2]], 
            'b--', alpha=0.5)
    ax.plot([v2[0], v_sum[0]], [v2[1], v_sum[1]], [v2[2], v_sum[2]], 
            'g--', alpha=0.5)
    
    # Formatting
    ax.set_xlim([0, 4])
    ax.set_ylim([0, 5])
    ax.set_zlim([0, 5])
    ax.set_xlabel('X axis', fontsize=11)
    ax.set_ylabel('Y axis', fontsize=11)
    ax.set_zlabel('Z axis', fontsize=11)
    ax.set_title('3D Vector Addition', fontsize=14, fontweight='bold')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Print 3D dot product
    dot_3d = np.dot(v1, v2)
    print(f"\n🔢 3D Dot Product: {v1} · {v2} = {dot_3d}")

plot_3d_vectors()
```

```python
# Cell 5: Animate sensor journey through feature space
from matplotlib.animation import FuncAnimation

def create_sensor_journey():
    """Animate sensor readings moving through 2D feature space"""
    
    # Generate synthetic sensor data (temperature vs current)
    np.random.seed(42)
    n_points = 50
    
    # Create a path through feature space (spiral pattern)
    t = np.linspace(0, 4*np.pi, n_points)
    temperature = 25 + 10 * np.sin(t) + np.random.normal(0, 0.5, n_points)
    current = 5 + 3 * np.cos(t * 0.8) + np.random.normal(0, 0.3, n_points)
    
    # Decision boundary: separates normal from anomalous
    # Line equation: current = slope * temperature + intercept
    slope = 0.2
    intercept = -3
    boundary_x = np.array([15, 40])
    boundary_y = slope * boundary_x + intercept
    
    # Classify points based on decision boundary
    classifications = []
    for temp, curr in zip(temperature, current):
        boundary_val = slope * temp + intercept
        if curr > boundary_val:
            classifications.append('normal')
        else:
            classifications.append('anomaly')
    
    # Create figure
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Plot decision boundary
    ax.plot(boundary_x, boundary_y, 'g-', linewidth=2, 
            label='Decision Boundary: current = 0.2×temp - 3')
    
    # Color map for points
    colors = {'normal': 'blue', 'anomaly': 'red'}
    
    # Scatter all points
    for classification in ['normal', 'anomaly']:
        mask = np.array(classifications) == classification
        ax.scatter(temperature[mask], current[mask], 
                   c=colors[classification], alpha=0.3, s=50,
                   label=f'{classification.capitalize()} points')
    
    # Animation: show the journey
    line, = ax.plot([], [], 'ro-', linewidth=2, markersize=8, label='Sensor journey')
    point, = ax.plot([], [], 'ro', markersize=10, markeredgecolor='black')
    
    def init():
        line.set_data([], [])
        point.set_data([], [])
        return line, point
    
    def animate(i):
        # Show path up to current point
        line.set_data(temperature[:i+1], current[:i+1])
        # Show current point
        point.set_data([temperature[i]], [current[i]])
        
        # Update title with current classification
        current_class = classifications[i]
        color = colors[current_class]
        ax.set_title(f'Sensor Journey - Step {i+1}/{n_points}\n'
                    f'Current reading: ({temperature[i]:.1f}°C, {current[i]:.1f}A) - '
                    f'{current_class.upper()}', 
                    fontsize=12, color=color)
        return line, point
    
    # Formatting
    ax.set_xlim(15, 42)
    ax.set_ylim(0, 8)
    ax.set_xlabel('Temperature (°C)', fontsize=12)
    ax.set_ylabel('Current (A)', fontsize=12)
    ax.set_title('Sensor Journey Through Feature Space', fontsize=14, fontweight='bold')
    ax.grid(True, alpha=0.3)
    ax.legend(loc='upper left')
    
    # Create animation
    anim = FuncAnimation(fig, animate, init_func=init, 
                         frames=n_points, interval=200, blit=True, repeat=True)
    
    plt.tight_layout()
    plt.show()
    
    # Also display as HTML for better viewing in notebook
    from IPython.display import HTML
    return HTML(anim.to_jshtml())

# Run the animation
print("🎬 Creating sensor journey animation...")
print("   Green line = Decision boundary (separates normal from anomalous)")
print("   Blue points = Normal readings")
print("   Red points = Anomalous readings")
print("   Red line = Sensor path through time")
print("\n⏳ Processing animation...")

# Uncomment to run animation (may take a moment)
# create_sensor_journey()

# Alternative: Static version without animation
def static_sensor_journey():
    """Static version for quick viewing"""
    np.random.seed(42)
    n_points = 50
    t = np.linspace(0, 4*np.pi, n_points)
    temperature = 25 + 10 * np.sin(t) + np.random.normal(0, 0.5, n_points)
    current = 5 + 3 * np.cos(t * 0.8) + np.random.normal(0, 0.3, n_points)
    
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Decision boundary
    boundary_x = np.array([15, 40])
    boundary_y = 0.2 * boundary_x - 3
    ax.plot(boundary_x, boundary_y, 'g-', linewidth=2, label='Decision Boundary')
    
    # Path with color gradient
    for i in range(n_points-1):
        ax.plot(temperature[i:i+2], current[i:i+2], 'r-', alpha=0.5, linewidth=1.5)
    
    # Scatter with classification
    classifications = current > (0.2 * temperature - 3)
    ax.scatter(temperature[classifications], current[classifications], 
               c='blue', s=50, alpha=0.7, label='Normal')
    ax.scatter(temperature[~classifications], current[~classifications], 
               c='red', s=50, alpha=0.7, label='Anomaly')
    
    ax.set_xlabel('Temperature (°C)', fontsize=12)
    ax.set_ylabel('Current (A)', fontsize=12)
    ax.set_title('Sensor Journey: 50 Readings Through Feature Space', fontsize=14, fontweight='bold')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

static_sensor_journey()
```

```python
# Cell 6: Explain how decision boundary works
def explain_decision_boundary():
    """Explain the mathematics of decision boundaries"""
    
    print("\n" + "="*70)
    print("🧠 UNDERSTANDING DECISION BOUNDARIES")
    print("="*70)
    
    print("""
    A decision boundary is a line (or hyperplane) that separates two classes.
    
    Equation: w·x + b = 0
    
    Where:
    • w = weight vector (perpendicular to the boundary)
    • x = feature vector [temperature, current]
    • b = bias term
    
    In our example:
    • Boundary: current = 0.2 × temperature - 3
    • Rearranged: -0.2 × temperature + 1 × current + 3 = 0
    • Weight vector w = [-0.2, 1]
    • Bias b = 3
    
    Classification rule:
    • If w·x + b > 0  → Normal reading
    • If w·x + b < 0  → Anomalous reading
    
    💡 Why this matters: Real ML models learn w and b from data!
    """)
    
    # Visualise weight vector
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Decision boundary
    x_range = np.array([15, 40])
    y_range = 0.2 * x_range - 3
    ax.plot(x_range, y_range, 'g-', linewidth=2, label='Decision Boundary')
    
    # Weight vector (perpendicular to boundary)
    weight = np.array([-0.2, 1])
    weight = weight / np.linalg.norm(weight) * 5  # Normalize and scale
    ax.quiver(25, 2, weight[0], weight[1], angles='xy', scale_units='xy',
              scale=1, color='purple', width=0.02, 
              label='Weight vector w (perpendicular to boundary)')
    
    ax.set_xlim(15, 42)
    ax.set_ylim(0, 8)
    ax.set_xlabel('Temperature (°C)')
    ax.set_ylabel('Current (A)')
    ax.set_title('Decision Boundary with Weight Vector', fontsize=14, fontweight='bold')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

explain_decision_boundary()
```
