```python
experiment_tracker = {}


def log_run(run_id, params, metrics):
    experiment_tracker[run_id] = {
        'params': params,
        'metrics': metrics
    }

def get_best_run(metric_name):
    best_run = None
    best_value = float('-inf')

    for run_id, data in experiment_tracker.items():
        value = data['metrics'].get(metric_name, None)
        
        if value is not None and value > best_value:
            best_value = value
            best_run = run_id

    return best_run


def summarise():
    print("------ Experiment Summary ------")
    
    for run_id, data in experiment_tracker.items():
        print(f"Run ID: {run_id}")
        
        print("  Params:")
        for k, v in data['params'].items():
            print(f"    {k}: {v}")
        
        print("  Metrics:")
        for k, v in data['metrics'].items():
            print(f"    {k}: {v}")
        
        print("-------------------------------")


log_run("run_1", 
        {'lr': 0.01, 'batch_size': 32}, 
        {'accuracy': 0.85, 'loss': 0.35})

log_run("run_2", 
        {'lr': 0.001, 'batch_size': 64}, 
        {'accuracy': 0.90, 'loss': 0.25})

log_run("run_3", 
        {'lr': 0.0001, 'batch_size': 32}, 
        {'accuracy': 0.88, 'loss': 0.30})

log_run("run_4", 
        {'lr': 0.005, 'batch_size': 16}, 
        {'accuracy': 0.92, 'loss': 0.20})



best = get_best_run('accuracy')
print("Best run (by accuracy):", best)

summarise()
```
