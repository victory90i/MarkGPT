```python
# Create a program that assigns letter grades
score = 85

if score >= 90:
    grade = "A"
    if score >= 97:
        grade += "+"
    elif score >= 93:
        grade += ""
    else:
        grade += "-"

elif score >= 80:
    grade = "B"
    if score >= 87:
        grade += "+"
    elif score >= 83:
        grade += ""
    else:
        grade += "-"

elif score >= 70:
    grade = "C"
    if score >= 77:
        grade += "+"
    elif score >= 73:
        grade += ""
    else:
        grade += "-"

elif score >= 60:
    grade = "D"
    if score >= 67:
        grade += "+"
    elif score >= 63:
        grade += ""
    else:
        grade += "-"

else:
    grade = "F"

# Print result
print("Score:", score)
print("Grade:", grade)
```

```python
# Test years
years = [2000, 1900, 2024, 2023]

for year in years:
    # Check leap year condition
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        result = "Leap Year"
    else:
        result = "Not a Leap Year"
    
    print(year, ":", result)
```

```python
for num in range(2, 51):   # Start from 2 (first prime)
    is_prime = True

    for i in range(2, num):   # Check divisibility
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num)
```
