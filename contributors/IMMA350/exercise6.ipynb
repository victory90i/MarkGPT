```python

val_losses = [0.45, 0.38, 0.34, 0.34, 0.34, 0.34, 0.33]

patience = 3

best_loss = float('inf')   
counter = 0               
epoch = 0                  
while epoch < len(val_losses):
    current_loss = val_losses[epoch]
    
    if current_loss < best_loss:
        best_loss = current_loss
        counter = 0 
    else:
        counter += 1 
    
    if counter == patience:
        print(f"Early stopping triggered at epoch {epoch + 1}")
        break
    
    epoch += 1
```
