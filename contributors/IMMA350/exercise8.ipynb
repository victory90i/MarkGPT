```python

raw_scores = [15, -3, 200, 42, 0, 99, -7, 55, 188, 30]


clipped_scores = [min(max(x, 0), 100) for x in raw_scores]


normalised = [x / 100 for x in clipped_scores]


labels = ['high' if x >= 0.6 else 'low' for x in normalised]


print("Clipped Scores:", clipped_scores)
print("Normalised:", normalised)
print("Labels:", labels)
```
