```python
def normalise_celsius_list(celsius_list):
    """
    Convert a list of Celsius temperatures to Kelvin and apply min-max normalisation.
    Returns the normalised values.
    """

    kelvin_list = [c + 273.15 for c in celsius_list]
    
 
    min_k = min(kelvin_list)
    max_k = max(kelvin_list)
    

    normalised_list = [(k - min_k) / (max_k - min_k) for k in kelvin_list]
    
    return normalised_list

temperatures_c = [10, 25, 37, 100, -5]
normalised_temperatures = normalise_celsius_list(temperatures_c)

print("Original temperatures (C):", temperatures_c)
print("Normalised temperatures:", normalised_temperatures)
```
