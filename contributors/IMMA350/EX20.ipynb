```python
def load_data():
    return [
        {'value': 10, 'label': 1},
        {'value': 25, 'label': 0},
        {'value': None, 'label': 1},
        {'value': -5, 'label': 0},
        {'value': 40, 'label': 1},
        {'value': 15, 'label': 0},
        {'value': 30, 'label': 1}
    ]

def clean_data(data):
    cleaned = []
    for item in data:
        if item['value'] is not None and item['value'] >= 0:
            cleaned.append(item)
    return cleaned


def normalise_data(data):
    values = [item['value'] for item in data]
    min_val = min(values)
    max_val = max(values)

    for item in data:
        item['value'] = (item['value'] - min_val) / (max_val - min_val)

    return data

def split_data(data, test_ratio=0.2):
    split_index = int(len(data) * (1 - test_ratio))
    train = data[:split_index]
    test = data[split_index:]
    return train, test

def run_pipeline():
    data = load_data()
    data = clean_data(data)
    data = normalise_data(data)
    train, test = split_data(data)

    print("Train size:", len(train))
    print("Test size:", len(test))
    print("First train sample:", train[0] if train else None)
    print("First test sample:", test[0] if test else None)

run_pipeline()
```
