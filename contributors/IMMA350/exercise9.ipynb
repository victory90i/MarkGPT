```python
def get_batches(data, batch_size):
    for i in range(0, len(data), batch_size):
        yield data[i:i + batch_size]



data = [f"sample_{i}" for i in range(1, 11)]



print("Batch size = 3")
for batch_num, batch in enumerate(get_batches(data, 3), start=1):
    print(f"Batch {batch_num}: {batch}")


print("\n" + "-"*40 + "\n")



print("Batch size = 4")
for batch_num, batch in enumerate(get_batches(data, 4), start=1):
    print(f"Batch {batch_num}: {batch}")
```
