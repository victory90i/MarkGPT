```python
best_val_accuracy = 0.0
checkpoint_log = []

def update_checkpoint(epoch, val_acc, model_name):
    global best_val_accuracy
    global checkpoint_log

    # Check if this is the best model so far
    is_best = False

    if val_acc > best_val_accuracy:
        best_val_accuracy = val_acc
        is_best = True

    # Save checkpoint info
    checkpoint_log.append({
        "epoch": epoch,
        "val_accuracy": val_acc,
        "model_name": model_name,
        "is_best": is_best
    })


# Simulated training results
update_checkpoint(1, 0.70, "model_A")
update_checkpoint(2, 0.75, "model_A")
update_checkpoint(3, 0.72, "model_A")
update_checkpoint(4, 0.80, "model_A")
update_checkpoint(5, 0.78, "model_A")

# Print results
print("Best Validation Accuracy:", best_val_accuracy)
print("\nCheckpoint Log:")
for entry in checkpoint_log:
    print(entry)
```
