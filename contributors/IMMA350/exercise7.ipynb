```python
def validate_sample(age, income, credit_score):
    if age < 0 or income < 0 or credit_score < 0:
        return 'Invalid: negative values not allowed'
    
    elif age < 18 or age > 100:
        return 'Invalid: age out of range'
    
    elif income > 1_000_000:
        return 'Flag: unusually high income — review'
    
    # Otherwise valid
    else:
        return 'Valid sample'


# Test cases
print(validate_sample(25, 50000, 700))      
print(validate_sample(-5, 40000, 650))        
print(validate_sample(120, 30000, 600))      
print(validate_sample(35, 2000000, 750))     
```
