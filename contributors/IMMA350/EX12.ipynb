```python

sample = [28, 'engineer', 85000, True, 720]
columns = ['age', 'occupation', 'income', 'employed', 'credit_score']


print("Original dataset row:")
for col, val in zip(columns, sample):
    print(f"{col}: {val}")


sample.append(False)
columns.append('loan_approved')

print("\nAfter adding 'loan_approved':")
for col, val in zip(columns, sample):
    print(f"{col}: {val}")


sample.remove(True)  # remove the value True corresponding to 'employed'
columns.remove('employed')

print("\nAfter dropping 'employed' feature:")
for col, val in zip(columns, sample):
    print(f"{col}: {val}")
```
