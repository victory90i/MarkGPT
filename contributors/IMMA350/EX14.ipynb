```python
def report_metrics(model_name, accuracy, loss, verbose=True, round_to=4):
    """
    Prints a formatted evaluation report.
    """

    accuracy = round(accuracy, round_to)
    loss = round(loss, round_to)

    if verbose:
        print("----- Evaluation Report -----")
        print(f"Model: {model_name}")
        print(f"Accuracy: {accuracy}")
        print(f"Loss: {loss}")
        print("-----------------------------\n")
    else:
        print(f"{model_name} | Accuracy: {accuracy}\n")


report_metrics("Model_A", 0.923456, 0.123456, True, 3)


report_metrics("Model_B", 0.856789, 0.234567)


report_metrics("Model_C", 0.912345, 0.198765, verbose=False)


report_metrics("Model_D", 0.987654, 0.111111, round_to=2)
```
