```python
import numpy as np

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline

# -----------------------------------
# Generate motor dataset
# -----------------------------------
np.random.seed(42)

n = 300

hours_op = np.random.uniform(0, 10, n)
ambient = np.random.uniform(20, 40, n)
load_pct = np.random.uniform(0.2, 1.0, n)
voltage = np.random.uniform(210, 240, n)

# Feature matrix
X = np.column_stack((hours_op, ambient, load_pct, voltage))

# Noise
noise = np.random.normal(0, 2, n)

# Target
y = (
    0.8 * hours_op +
    0.5 * ambient +
    25 * load_pct +
    0.05 * voltage +
    noise
)

# -----------------------------------
# (a) Linear Regression with 5-fold CV
# -----------------------------------
linear_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LinearRegression())
])

linear_scores = cross_val_score(
    linear_pipeline,
    X,
    y,
    cv=5,
    scoring="r2"
)

# -----------------------------------
# (b) Print fold scores and statistics
# -----------------------------------
print("Linear Regression R2 scores:")

for i, score in enumerate(linear_scores, start=1):
    print(f"Fold {i}: {score:.4f}")

print("\nLinear Regression Mean R2:")
print(f"{linear_scores.mean():.4f} ± {linear_scores.std():.4f}")

# -----------------------------------
# (c) Polynomial Degree 2 Pipeline
# -----------------------------------
poly_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("poly", PolynomialFeatures(degree=2)),
    ("model", LinearRegression())
])

poly_scores = cross_val_score(
    poly_pipeline,
    X,
    y,
    cv=5,
    scoring="r2"
)

print("\nPolynomial Degree 2 R2 scores:")

for i, score in enumerate(poly_scores, start=1):
    print(f"Fold {i}: {score:.4f}")

print("\nPolynomial Degree 2 Mean R2:")
print(f"{poly_scores.mean():.4f} ± {poly_scores.std():.4f}")
```
