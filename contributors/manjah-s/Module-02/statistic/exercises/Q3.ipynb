```python
import pandas as pd
import numpy as np

# Create table
df = pd.DataFrame({
    "name": ['Alice','Bob','Carlos','Diana','Eve'],
    "score": [88, 72, 95, 61, 83],
    "passed": [True,True,True,False,True]
})

# Step 1: Add grade column
conditions = [
    df['score'] >= 90,
    df['score'] >= 80
]

choices = ['A', 'B']

df['grade'] = np.select(conditions, choices, default='C')

# Step 2: Filter passing students
passing = df[df['passed']]

# Step 3: Summary statistics
summary = df['score'].describe()

print(df)
print("\nPassing students:\n", passing)
print("\nSummary:\n", summary)
```
