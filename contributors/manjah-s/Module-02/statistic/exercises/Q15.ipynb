```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

from scipy.stats import skew

data = np.random.lognormal(mean=3, sigma=1, size=800)

log_data = np.log1p(data)

print("Original skew:", skew(data))
print("Log skew:", skew(log_data))

fig, ax = plt.subplots(1,2, figsize=(12,4))

sns.histplot(data, kde=True, ax=ax[0])
ax[0].set_title("Original")

sns.histplot(log_data, kde=True, ax=ax[1])
ax[1].set_title("Log Transformed")

plt.show()
```
