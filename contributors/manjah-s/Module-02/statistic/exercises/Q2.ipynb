```python
import numpy as np

# Step 1: Create numbers from 0 to 4π
x = np.linspace(0, 4*np.pi, 20)

# Step 2: Apply sine to every value
y = np.sin(x)

# Step 3a: index of max value
max_index = np.argmax(y)

# Step 3b: mean and std
mean_val = np.mean(y)
std_val = np.std(y)

# Step 3c: count values > 0.5
count_above = np.sum(y > 0.5)

print("Max index:", max_index)
print("Mean:", mean_val)
print("Std:", std_val)
print("Count > 0.5:", count_above)
```
