```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Generate data
data = np.random.lognormal(mean=5.0, sigma=0.5, size=28)

# Add outliers
data = np.append(data, [800, 1200])

# Compute statistics
mean_val = np.mean(data)
median_val = np.median(data)
mode_val = stats.mode(data).mode

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)

# Plot
sns.histplot(data, kde=True)

plt.axvline(mean_val, linestyle='--', label='Mean')
plt.axvline(median_val, linestyle='-', label='Median')
plt.axvline(mode_val, linestyle=':', label='Mode')

plt.legend()
plt.title("House Prices Distribution")
plt.show()
```
