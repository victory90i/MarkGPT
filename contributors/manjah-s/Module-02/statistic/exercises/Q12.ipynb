```python

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

data = np.array([12, 15, 14, 10, 18, 16, 11, 17, 13, 14])

# Mean
mean = np.mean(data)

# Deviations
deviations = data - mean

# Squared deviations
sq_devs = deviations ** 2

# Population variance
pop_var = np.mean(sq_devs)

# Sample variance
sample_var = np.sum(sq_devs) / (len(data) - 1)

# Verify
print("Manual Population:", pop_var)
print("NumPy Population:", np.var(data, ddof=0))

print("Manual Sample:", sample_var)
print("NumPy Sample:", np.var(data, ddof=1))
```
