```python
from scipy import stats
import numpy as np

datasets = {
    "Normal": np.random.normal(0,1,500),
    "Exponential": np.random.exponential(1,500),
    "Uniform": np.random.uniform(0,1,500)
}

print("Name | Skew | P-value | Result")

for name, data in datasets.items():
    stat, p = stats.normaltest(data)
    skew = stats.skew(data)
    
    result = "NORMAL" if p > 0.05 else "NOT NORMAL"
    
    print(name, "|", round(skew,2), "|", round(p,5), "|", result)
```
