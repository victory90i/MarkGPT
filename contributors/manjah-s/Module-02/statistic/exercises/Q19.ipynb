```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

data = np.random.normal(65,20,200)

scaled = (data - np.mean(data)) / np.std(data, ddof=1)

print("Mean:", round(np.mean(scaled),6))
print("Std:", round(np.std(scaled, ddof=1),6))

plt.hist(data, alpha=0.5, label='Original')
plt.hist(scaled, alpha=0.5, label='Scaled')

plt.legend()
plt.show()
```
