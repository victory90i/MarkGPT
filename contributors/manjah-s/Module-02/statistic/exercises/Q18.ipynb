```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

temp = np.cumsum(np.random.normal(0,2,365)) + 15
s = pd.Series(temp)

plt.plot(s, alpha=0.5, label='Raw')
plt.plot(s.rolling(7).mean(), label='7-day')
plt.plot(s.rolling(30).mean(), linestyle='--', label='30-day')

plt.legend()
plt.title("Temperature Trends")
plt.show()
```
