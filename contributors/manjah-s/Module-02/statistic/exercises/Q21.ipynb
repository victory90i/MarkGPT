```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

cities = ['Lagos','Douala','Nairobi','Accra']

df = pd.DataFrame({
    "city": np.random.choice(cities,300),
    "age": np.random.randint(22,60,300)
})

income_map = {
    "Lagos": np.random.lognormal(5,1),
    "Douala": np.random.lognormal(4.5,1),
    "Nairobi": np.random.lognormal(4.8,1),
    "Accra": np.random.lognormal(5.2,1)
}

df['income'] = df['city'].map(income_map)

print(df.groupby('city')['income'].agg(['mean','median','std','count']))
```
