```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

data = {'Room A':35, 'Room B':25, 'Room C':40}

y_offset = 0

for room, count in data.items():
    symbols = count // 5
    
    for i in range(symbols):
        x = i % 10
        y = y_offset - (i // 10)
        ax.text(x, y, '★', fontsize=12)
    
    ax.text(-2, y_offset, room)
    y_offset -= 3

ax.set_xlim(-3,10)
ax.set_ylim(-10,2)
ax.axis('off')

plt.show()
```
