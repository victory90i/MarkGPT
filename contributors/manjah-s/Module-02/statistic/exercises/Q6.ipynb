```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 100)

fig, ax = plt.subplots()

ax.plot(x, x**2, linestyle='-', label='x^2')
ax.plot(x, x**3, linestyle='--', label='x^3')

ax.axhline(0, linestyle='--')
ax.axvline(0, linestyle='--')

ax.set_title("Polynomial Graph", fontweight='bold')
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")

ax.grid(alpha=0.3)
ax.legend()

plt.show()
```
