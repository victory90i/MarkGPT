```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

df = pd.DataFrame({
    "income": np.random.lognormal(5,1,100),
    "age": np.random.normal(38,10,100),
    "spend": np.random.exponential(5,100),
    "credit_score": np.random.normal(700,80,100)
})

summary = {}

for col in df.columns:
    summary[col] = [
        df[col].min(),
        np.percentile(df[col],25),
        np.median(df[col]),
        np.percentile(df[col],75),
        df[col].max()
    ]

summary_df = pd.DataFrame(summary, index=['min','Q1','median','Q3','max'])
print(summary_df.to_string())
```
