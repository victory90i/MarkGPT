#Standard deviation


```python
import pandas as pd

data = np.array([10, 12, 14, 16, 20])

mean = np.mean(data)

deviation = data - mean
sq_devs = deviation**2
variance_pop = np.mean(sq_devs)
std_pop = np.sqrt(variance_pop)

print(f"population variance: {np.var(data, ddof=0): .4f}")
print(f"sample variance: {np.var(data, ddof=1): .4f}")
print(f"std deviation (pop): {np.std(data, ddof=0): .4f}")
print(f"std deviation (smp): {np.std(data, ddof=1): .4f}")
```

Range, IQR and Percentiles

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

data = np.array([2, 5, 7, 9, 10, 12, 15, 18, 20, 100])

print(f"Range: {np.max(data) - np.min(data)}")

Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

print(f'Q1 ={Q1}, Q2 ={Q2}, Q3 ={Q3}')
print(f'IQR = {IQR}')

lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
outliers = data[(data < lower_fence) | (data >  upper_fence)]
print(f'outliers datected: {outliers}')
```
