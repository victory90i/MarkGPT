```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# Reproducibility
np.random.seed(42)

# Deck setup
n_draws = 200000

# Represent cards:
# Hearts = 1 suit out of 4
# Face cards = J, Q, K = 3 ranks out of 13

# Simulate suits (0–3), hearts = 0
suits = np.random.randint(0, 4, n_draws)

# Simulate ranks (1–13), face cards = 11,12,13
ranks = np.random.randint(1, 14, n_draws)

# Events
is_heart = suits == 0
is_face = np.isin(ranks, [11, 12, 13])

# Probabilities
P_heart = np.mean(is_heart)
P_face = np.mean(is_face)
P_heart_and_face = np.mean(is_heart & is_face)
P_heart_or_face = np.mean(is_heart | is_face)

# Addition rule check
P_addition = P_heart + P_face - P_heart_and_face

# Print results
print("Probabilities from simulation:")
print(f"P(heart) = {P_heart:.5f}")
print(f"P(face card) = {P_face:.5f}")
print(f"P(heart AND face) = {P_heart_and_face:.5f}")
print(f"P(heart OR face) = {P_heart_or_face:.5f}")

print("\nAddition Rule Check:")
print(f"P(A) + P(B) - P(A AND B) = {P_addition:.5f}")

# Regions for Venn diagram
heart_only = P_heart - P_heart_and_face
face_only = P_face - P_heart_and_face
intersection = P_heart_and_face

# Plot Venn-style diagram
fig, ax = plt.subplots()

# Draw circles (no explicit color set)
circle1 = Circle((0.4, 0.5), 0.3, fill=False)
circle2 = Circle((0.6, 0.5), 0.3, fill=False)

ax.add_patch(circle1)
ax.add_patch(circle2)

# Annotate regions
ax.text(0.3, 0.5, f"{heart_only:.3f}", ha='center')
ax.text(0.7, 0.5, f"{face_only:.3f}", ha='center')
ax.text(0.5, 0.5, f"{intersection:.3f}", ha='center')

# Labelsimport numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# Reproducibility
np.random.seed(42)

# Deck setup
n_draws = 200_000

# Represent cards:
# Hearts = 1 suit out of 4
# Face cards = J, Q, K = 3 ranks out of 13

# Simulate suits (0–3), hearts = 0
suits = np.random.randint(0, 4, n_draws)

# Simulate ranks (1–13), face cards = 11,12,13
ranks = np.random.randint(1, 14, n_draws)

# Events
is_heart = suits == 0
is_face = np.isin(ranks, [11, 12, 13])

# Probabilities
P_heart = np.mean(is_heart)
P_face = np.mean(is_face)
P_heart_and_face = np.mean(is_heart & is_face)
P_heart_or_face = np.mean(is_heart | is_face)

# Addition rule check
P_addition = P_heart + P_face - P_heart_and_face

# Print results
print("Probabilities from simulation:")
print(f"P(heart) = {P_heart:.5f}")
print(f"P(face card) = {P_face:.5f}")
print(f"P(heart AND face) = {P_heart_and_face:.5f}")
print(f"P(heart OR face) = {P_heart_or_face:.5f}")

print("\nAddition Rule Check:")
print(f"P(A) + P(B) - P(A AND B) = {P_addition:.5f}")

# Regions for Venn diagram
heart_only = P_heart - P_heart_and_face
face_only = P_face - P_heart_and_face
intersection = P_heart_and_face

# Plot Venn-style diagram
fig, ax = plt.subplots()

# Draw circles (no explicit color set)
circle1 = Circle((0.4, 0.5), 0.3, fill=False)
circle2 = Circle((0.6, 0.5), 0.3, fill=False)

ax.add_patch(circle1)
ax.add_patch(circle2)

# Annotate regions
ax.text(0.3, 0.5, f"{heart_only:.3f}", ha='center')
ax.text(0.7, 0.5, f"{face_only:.3f}", ha='center')
ax.text(0.5, 0.5, f"{intersection:.3f}", ha='center')

# Labels
ax.text(0.3, 0.8, "Heart", ha='center')
ax.text(0.7, 0.8, "Face Card", ha='center')

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')

plt.title("Venn Diagram of Card Probabilities")
plt.show()
ax.text(0.3, 0.8, "Heart", ha='center')
ax.text(0.7, 0.8, "Face Card", ha='center')

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')

plt.title("Venn Diagram of Card Probabilities")
plt.show()
```
