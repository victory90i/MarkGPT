```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate 50,000 die rolls
rolls = np.random.randint(1, 7, 50000)

# Estimated probabilities
p_4_est = np.mean(rolls == 4)
p_even_est = np.mean(rolls % 2 == 0)
p_above4_est = np.mean(rolls > 4)

# Theoretical probabilities
p_4_theo = 1/6
p_even_theo = 3/6
p_above4_theo = 2/6

# Print results
print("Estimated vs Theoretical Probabilities:")
print(f"(a) P(rolling a 4): {p_4_est:.4f} vs {p_4_theo:.4f}")
print(f"(b) P(even number): {p_even_est:.4f} vs {p_even_theo:.4f}")
print(f"(c) P(above 4): {p_above4_est:.4f} vs {p_above4_theo:.4f}")

# Compute probabilities for each face
faces = np.arange(1, 7)
est_probs = [np.mean(rolls == i) for i in faces]
theo_probs = [1/6] * 6

# Plot bar chart
x = np.arange(len(faces))

plt.figure()
plt.bar(x - 0.2, est_probs, width=0.4, label="Estimated")
plt.bar(x + 0.2, theo_probs, width=0.4, label="Theoretical")

plt.xticks(x, faces)
plt.xlabel("Die Face")
plt.ylabel("Probability")
plt.title("Estimated vs Theoretical Probabilities for Die Faces")
plt.legend()

plt.show()
```
