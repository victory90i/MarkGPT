```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

n=  500000
prevalence =  0.02
sensitivity =0.94
specificity =0.88

disease = np.random.rand(n)<prevalence
test_positive = np.zeros(n, dtype=bool)
test_positive[disease] = np.random.rand(np.sum(disease))< sensitivity
healthy = ~disease
test_positive[healthy]= np.random.rand(np.sum(healthy))<(1-specificity)
positive = test_positive
negative = ~test_positive


P_disease_given_pos = np.sum(disease & positive) / np.sum(positive)
P_disease_given_neg = np.sum(disease & negative) / np.sum(negative)

P_D = prevalence
P_notD = 1 - P_D

P_pos_given_D = sensitivity
P_pos_given_notD = 1 - specificity

P_pos = P_pos_given_D * P_D + P_pos_given_notD * P_notD

P_D_given_pos_bayes = (P_pos_given_D * P_D) / P_pos
P_D_given_neg_bayes = ((1 - sensitivity) * P_D) / (1 - P_pos)

# 5. Print formatted table
print("\n=== Results (Simulation vs Bayes) ===\n")
print(f"{'Metric':<30}{'Simulation':<15}{'Bayes':<15}")
print("-" * 60)
print(f"{'P(Disease | Positive)':<30}{P_disease_given_pos:<15.5f}{P_D_given_pos_bayes:<15.5f}")
print(f"{'P(Disease | Negative)':<30}{P_disease_given_neg:<15.5f}{P_D_given_neg_bayes:<15.5f}")

# Optional: also show base rates
print("\nAdditional Info:")
print(f"P(Positive test) ≈ {np.mean(positive):.5f}")
print(f"P(Disease overall) = {prevalence:.5f}")
```
