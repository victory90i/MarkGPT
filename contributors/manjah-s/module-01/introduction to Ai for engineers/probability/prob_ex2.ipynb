```python
import numpy as np
import matplotlib.pyplot as plt

# Number of flips
n = 10000

# Simulate coin flips (1 = heads, 0 = tails)
flips = np.random.randint(0, 2, n)

# Running proportion of heads
running_prop = np.cumsum(flips) / np.arange(1, n + 1)

# Theoretical probability
p = 0.5

# Standard error at each step
se = np.sqrt(0.25 / np.arange(1, n + 1))

# Create x values
x = np.arange(1, n + 1)

# Plot
plt.figure()
plt.plot(x, running_prop, color='navy', label='Running Proportion')
plt.axhline(p, linestyle='--', color='gold', label='P = 0.5')

# Shaded ±1 SE band
plt.fill_between(x, p - se, p + se, alpha=0.2)

# Logarithmic x-axis
plt.xscale('log')

# Labels and title
plt.xlabel("Number of Flips (log scale)")
plt.ylabel("Proportion of Heads")
plt.title("Running Proportion of Heads with ±1 SE Band")
plt.legend()

plt.show()

# Comment:
# Standard Error = sqrt(0.25 / n)
# Solve sqrt(0.25 / n) < 0.01:
# 0.25 / n < 0.0001
# n > 2500
# So, at least 2500 flips are needed for SE < 0.01
```
