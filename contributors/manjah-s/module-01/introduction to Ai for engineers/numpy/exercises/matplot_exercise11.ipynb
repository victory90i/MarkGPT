```python
import numpy as np
import matplotlib.pyplot as plt


A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])


elementwise = A * B


matrixprod = A @ B


fig, axes = plt.subplots(1, 2, figsize=(8, 4))


im1 = axes[0].imshow(elementwise, cmap='coolwarm')
axes[0].set_title("Element-wise Product (A * B)")
fig.colorbar(im1, ax=axes[0])


im2 = axes[1].imshow(matrixprod, cmap='coolwarm')
axes[1].set_title("Matrix Product (A @ B)")
fig.colorbar(im2, ax=axes[1])

plt.tight_layout()
plt.show()
```
