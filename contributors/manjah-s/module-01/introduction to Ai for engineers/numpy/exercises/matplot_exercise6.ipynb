```python
import matplotlib.pyplot as plt
import numpy as np


a = np.array([3, 1])
b = np.array([1, 3])
c = a + b

plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='navy')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='gold')
plt.quiver(0, 0, c[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', linestyle='--', label='c = a + b')


plt.xlim(0, 5)
plt.ylim(0, 5)
plt.gca().set_aspect('equal')  

plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Vectors a and b')

plt.show()
```
