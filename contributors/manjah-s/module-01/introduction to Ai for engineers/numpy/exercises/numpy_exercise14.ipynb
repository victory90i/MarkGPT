```python
import numpy as np


A = np.array([[2, 0],
              [0, 3]])
det_A = np.linalg.det(A)


theta = np.deg2rad(60)
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])
det_R = np.linalg.det(R)


S = np.array([[1, 2],
              [1, 2]])
det_S = np.linalg.det(S)

print("Determinant of scaling matrix:", det_A)
print("Interpretation: scales area by factor of 6.")

print("Determinant of rotation matrix:", det_R)
print("Interpretation: preserves area and orientation (rigid motion).")

print("Determinant of singular matrix:", det_S)
print("Interpretation: collapses space into lower dimension, not invertible.")
```
