```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 10×10 grid of points
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
points = np.vstack([X.ravel(), Y.ravel()])  # shape (2, 100)


S = np.array([[2, 0],
              [0, 0.5]])


transformed = S @ points


fig, axes = plt.subplots(1, 2, figsize=(10, 5))


axes[0].scatter(points[0], points[1], color='navy')
axes[0].set_title("Original Grid")
axes[0].set_aspect('equal')
axes[0].grid(True)

axes[1].scatter(transformed[0], transformed[1], color='crimson')
axes[1].set_title("Transformed Grid (Scaling)")
axes[1].set_aspect('equal')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
