```python
import numpy as np


A = np.array([[2, 1],
              [5, 3]])


A_inv = np.linalg.inv(A)
print("Inverse of A:\n", A_inv)


identity_check = np.round(A @ A_inv, 10)
print("\nA @ A_inv (rounded):\n", identity_check)


b = np.array([3, 7])


x_via_inv = A_inv @ b
print("\nSolution using A_inv:", x_via_inv)


x_via_solve = np.linalg.solve(A, b)
print("Solution using np.linalg.solve:", x_via_solve)
```
