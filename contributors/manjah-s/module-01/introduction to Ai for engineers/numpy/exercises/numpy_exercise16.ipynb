```python
import numpy as np

# Shear transformation matrix
A = np.array([[1, 1],
              [0, 1]])

# Reflection matrix (reflection across the y-axis)
B = np.array([[-1, 0],
              [ 0, 1]])

# Compute determinants
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)

print("Shear matrix A:\n", A)
print("Determinant of A:", det_A)

print("\nReflection matrix B:\n", B)
print("Determinant of B:", det_B)
```
