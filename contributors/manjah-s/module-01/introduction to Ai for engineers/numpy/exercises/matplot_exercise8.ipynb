```python
import numpy as np
import matplotlib.pyplot as plt


vectors = np.random.randn(5, 3)


norms = np.linalg.norm(vectors, axis=1)


labels = [f'V{i+1}' for i in range(5)]


fig, ax = plt.subplots()
bars = ax.bar(labels, norms, color='skyblue')


mean_norm = norms.mean()
ax.axhline(mean_norm, color='red', linestyle='dashed', label=f'Mean = {mean_norm:.2f}')

ax.set_ylabel('Magnitude (Norm)')
ax.set_title('Magnitudes of Random 3-D Vectors')
ax.legend()

for bar, val in zip(bars, norms):
    ax.text(bar.get_x() + bar.get_width()/2, val + 0.05, f'{val:.2f}',
            ha='center', va='bottom')

plt.show()
```
