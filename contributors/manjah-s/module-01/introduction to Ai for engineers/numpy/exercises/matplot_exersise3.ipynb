```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0, 6)
y1= x
y2 = x**2

plt.figure(figsize=(8, 4))  

plt.plot(x, y1, color='blue', label='y =x' )
plt.plot(x, y2, color='red', label='y = x^2')

plt.title("y=x and y=x^2")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.grid(True)
plt.legend()

plt.show()
```
