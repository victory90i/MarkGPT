```python
import matplotlib.pyplot as plt
import numpy as np

# Define 
a = np.array([3, 1])
b = np.array([1, 3])

# Plot vector a (navy) and b (gold), both starting at origin (0,0)
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='navy')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='gold')

# Set axis limits so arrows are not distorted
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.gca().set_aspect('equal')  # Equal scaling for x and y

# Add grid and labels for clarity
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Vectors a and b')

plt.show()

```
