```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 360)


x = np.cos(theta)
y = np.sin(theta)


fig, ax = plt.subplots()
ax.plot(x, y, 'k')  


angles = [0, np.pi/3, 2*np.pi/3]  
colors = ['red', 'blue', 'green']


for ang, col in zip(angles, colors):
    ax.quiver(0, 0, np.cos(ang), np.sin(ang),
              angles='xy', scale_units='xy', scale=1,
              color=col, label=f'{np.degrees(ang):.4f}°')


ax.set_aspect('equal')
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)


ax.grid(True)
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Unit Circle with Unit Vectors')

plt.show()
```
