```python

def confusion_matrix(y_true, y_pred, labels):
    
    cm = {true_label: {pred_label: 0 for pred_label in labels} 
          for true_label in labels}
    for true, pred in zip(y_true, y_pred):
        if true in cm and pred in cm[true]:
            cm[true][pred] += 1
    
    return cm

def pretty_print_cm(cm, labels):
    
    print("\nConfusion Matrix:")
    
    header = " " * 12 + "Predicted →"
    for label in labels:
        header += f"{label:>10}"
    print(header)
    
    print(" " * 11 + "-" * (len(labels) * 10 + 5))
    
    for true_label in labels:
        row = f"True {true_label:>6}: "
        for pred_label in labels:
            count = cm[true_label][pred_label]
            row += f"{count:>10}"
        print(row)
    


def calculate_metrics(cm, labels):
    metrics = {}
    
    for label in labels:
        
        tp = cm[label][label]
        
        fp = sum(cm[true_label][label] for true_label in labels if true_label != label)
    
        fn = sum(cm[label][pred_label] for pred_label in labels if pred_label != label)
        
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0
        
        metrics[label] = {
            'precision': precision,
            'recall': recall,
            'tp': tp,
            'fp': fp,
            'fn': fn
        }
    
    return metrics

def print_metrics(metrics, labels):

    print("\nPer-Class Metrics:")
    
    for label in labels:
        prec = metrics[label]['precision']
        rec = metrics[label]['recall']
        print(f"{label:>6} - Precision: {prec:.3f}, Recall: {rec:.3f}")
    

    avg_precision = sum(metrics[label]['precision'] for label in labels) / len(labels)
    avg_recall = sum(metrics[label]['recall'] for label in labels) / len(labels)
    
    print("-" * 50)
    print(f"Macro Avg - Precision: {avg_precision:.3f}, Recall: {avg_recall:.3f}")


def test_confusion_matrix():
    
    print("\n" + "=" * 50)
    print("Q20: Confusion Matrix")
    print("=" * 50)
    
    labels = ['cat', 'dog', 'bird']
    
    y_true = [
        'cat', 'cat', 'cat', 'cat', 'cat',  
        'dog', 'dog', 'dog', 'dog', 'dog',  
        'bird', 'bird', 'bird', 'bird', 'bird',  
        'cat', 'dog', 'bird'  
    ]
    
    y_pred = [
        'cat', 'cat', 'dog', 'cat', 'cat',   
        'dog', 'cat', 'dog', 'bird', 'dog',  
        'bird', 'bird', 'cat', 'bird', 'dog',  
        'bird', 'dog', 'cat'  
    ]
    
    print(f"\nNumber of samples: {len(y_true)}")
    print(f"Sample predictions (first 10):")
    for i in range(min(10, len(y_true))):
        print(f"  Sample {i+1}: True={y_true[i]:>4}, Pred={y_pred[i]:>4}")
    
    cm = confusion_matrix(y_true, y_pred, labels)
    pretty_print_cm(cm, labels)
    metrics = calculate_metrics(cm, labels)
    print_metrics(metrics, labels)
test_confusion_matrix()
```
