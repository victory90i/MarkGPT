```python
best_val_accuracy =0.0
checkpoint_log = []

def update_checkpoint(epoch, val_acc, model_name):
    global best_val_accuracy 
    global checkpoint_log

    is_best = False

    if val_acc > best_val_accuracy:
        best_val_accuracy = val_acc
        is_best = True

    checkpoint ={
        "epoch": epoch,
        "model": model_name,
        "val_accuracy": val_acc,
        "is_best": is_best
    }
    checkpoint_log.append(checkpoint)

model_name = "my_model"

val_accuracies =[0.72, 0.75, 0.74, 0.78, 0.77]

for epoch, acc in enumerate(val_accuracies, start=1):
    update_checkpoint(epoch, acc, model_name)

for entry in checkpoint_log:
    print(entry)

print("Best validation accuracy:", best_val_accuracy)
```
