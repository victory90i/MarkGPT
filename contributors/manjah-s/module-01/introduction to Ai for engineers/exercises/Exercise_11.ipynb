```python
def celsuis_to_kelvin(c):
    return c + 273.15
```

```python

   def normalise(value, min_val, max_val):
    return (value - min_val)/(max_val - min_val)
     
```

```python

    

temp_celsuis =[10, 25, 37, 100, -5]
temp_in_kelvin = [ celsuis_to_kelvin(c) for c in temp_celsuis]
min_val = min(temp_in_kelvin)
max_val = max(temp_in_kelvin)
normalise_temp = [normalise(value, min_val, max_val) for value in temp_celsuis]


print(temp_in_kelvin, normalise_temp)
```
