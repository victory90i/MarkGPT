```python
features_source_a = {'age', 'income', 'credit_score', 'loan_amount', 'zip_code'}
features_source_b = {'age',  'income', 'employment_status', 'zip_code', 'region'}
```

```python
all_unique_features = features_source_a | features_source_b
print("all unique features: ", all_unique_features)
```

```python
common_features = features_source_a & features_source_b
print("features in both sources", common_features)
```

```python
exclusive_a = features_source_a - features_source_b
exclusive_b = features_source_b - features_source_a
print(f"exclusive to a {exclusive_a} and to b {exclusive_b}")
```
