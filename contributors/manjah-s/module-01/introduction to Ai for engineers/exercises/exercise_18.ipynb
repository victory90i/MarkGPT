```python

experiment_tracker = {}


def log_run(run_id, params, metrics):
    global experiment_tracker
    experiment_tracker[run_id] = {"params": params, "metrics": metrics}

def get_best_run(metric_name):
    global experiment_tracker
    best_run = None
    best_value = float('-inf')
    
    for run_id, info in experiment_tracker.items():
        value = info['metrics'].get(metric_name, float('-inf'))
        if value > best_value:
            best_value = value
            best_run = run_id
    
    return best_run

def summarise():
    global experiment_tracker
    print(f"{'Run ID':<10} {'Params':<40} {'Metrics'}")
    
    for run_id, info in experiment_tracker.items():
        print(f"{run_id:<10} {str(info['params']):<40} {info['metrics']}")


log_run("run_001", {"lr": 0.01, "batch": 32}, {"accuracy": 0.85, "loss": 0.35})
log_run("run_002", {"lr": 0.001, "batch": 64}, {"accuracy": 0.88, "loss": 0.30})
log_run("run_003", {"lr": 0.005, "batch": 32}, {"accuracy": 0.82, "loss": 0.40})
log_run("run_004", {"lr": 0.01, "batch": 64}, {"accuracy": 0.90, "loss": 0.28})


best_run = get_best_run("accuracy")
print("Best run for accuracy:", best_run)


print("\nAll runs summary:")
summarise()
```
