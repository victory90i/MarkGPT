```python

def load_data():
    return [
        {'value': 10, 'label': 'A'},
        {'value': 25, 'label': 'B'},
        {'value': None, 'label': 'C'},
        {'value': -5, 'label': 'D'},
        {'value': 30, 'label': 'E'},
        {'value': 15, 'label': 'F'},
        {'value': None, 'label': 'G'},
        {'value': 20, 'label': 'H'},
        {'value': -2, 'label': 'I'},
        {'value': 40, 'label': 'J'},
        {'value': 5, 'label': 'K'},
        {'value': 35, 'label': 'L'}
    ]

def clean_data(data):
    
    return [item for item in data if item['value'] is not None and item['value'] >= 0]

def normalise_data(data):
    
    if not data:
        return data
    
    values = [item['value'] for item in data]
    min_val = min(values)
    max_val = max(values)
    
    if min_val == max_val:
        for item in data:
            item['value'] = 0.0
    else:
        for item in data:
            item['value'] = (item['value'] - min_val) / (max_val - min_val)
    
    return data

def split_data(data, test_ratio=0.2):
    
    split_idx = int(len(data) * (1 - test_ratio))
    train = data[:split_idx]
    test = data[split_idx:]
    return (train, test)

def run_pipeline():
    
    print("=" * 50)
    print("Q19: Data Pipeline")
    
    
    data = load_data()
    print(f"Step 1 - Loaded data: {len(data)} records")
    print(f"  Sample: {data[:2]}")
    
    data = clean_data(data)
    print(f"\nStep 2 - After cleaning: {len(data)} records")
    print(f"  Sample: {data[:2]}")
    
    data = normalise_data(data)
    print(f"\nStep 3 - After normalisation:")
    print(f"  Sample: {data[:2]}")
    
    train, test = split_data(data, test_ratio=0.2)
    print(f"\nStep 4 - After splitting:")
    print(f"  Train set size: {len(train)} records")
    print(f"  Test set size: {len(test)} records")
    print(f"  Train first entry: {train[0] if train else 'None'}")
    print(f"  Test first entry: {test[0] if test else 'None'}")
    
    return train, test
train_data, test_data = run_pipeline()
```
