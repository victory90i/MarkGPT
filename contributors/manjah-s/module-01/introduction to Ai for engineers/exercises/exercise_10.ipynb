```python
def tokenize(text):
    text = text.lower()
    punctuation = ".,!?;:'\""
    for p in punctuation:
        text = text.replace(p, "")
    
    words = text.split()
    stop_words = ['a', 'an', 'the', 'is', 'in', 'of', 'and', 'to']
    
    tokens = []
    for word in words:
        if word not in stop_words:
            tokens.append(word)
    
    return tokens


text = "The quick brown fox jumps over the lazy dog and the cat."


tokens = tokenize(text)


vocab = sorted(list(set(tokens)))
freq = {}
for word in tokens:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

print("Tokens:", tokens)
print("Vocabulary:", vocab)
print("Frequencies:", freq)
```
