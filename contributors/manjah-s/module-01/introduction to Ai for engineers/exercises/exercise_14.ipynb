```python
def report_metrics(model_name, accuracy, loss, verbose=True, round_to = 4):
    accuracy = round(accuracy, round_to)
    loss = round(loss, round_to)

    if verbose:
        print(f"model: {model_name}")
        print(f"accuracy: {accuracy}")
        print(f"loss: {loss}")
    else:
        print(f"Model: {model_name} | accuracy: {accuracy}")

report_metrics("model_A", 0.94244848, 0.1838383, True, 3)
```

```python
report_metrics("model_A", 0.94244848, 0.18383)
```

```python
report_metrics("model_A", 0.94244848, 0.1838383, verbose=False)
```

```python
report_metrics("model_A", 0.94244848, 0.1838383, round_to =2)
```
