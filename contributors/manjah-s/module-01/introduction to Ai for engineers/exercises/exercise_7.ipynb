```python
def data_validation(age, income, credit_score):
    if age < 0 or income < 0 or credit_score < 0:
        print("negative values not allowed")

    elif age < 18 or age > 100:
        print("Invalide age, out of range")
    elif income > 1000000:
        print("usually high income")
    else:
        print("valid sample")
data_validation(18, 1000, 3)
```
