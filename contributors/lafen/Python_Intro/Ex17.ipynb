```python
features_source_a = {'age', 'income', 'credit_score', 'loan_amount', 'zip_code'}
features_source_b = {'age', 'income', 'employment_status', 'zip_code', 'region'}

# (1) All unique features
print("Union:", features_source_a | features_source_b)

# (2) Common features
print("Intersection:", features_source_a & features_source_b)

# (3) Exclusive to each
print("Only in A:", features_source_a - features_source_b)
print("Only in B:", features_source_b - features_source_a)

# (4) In one but not both
print("Symmetric Difference:", features_source_a ^ features_source_b)
```
