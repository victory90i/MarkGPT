```python
def confusion_matrix(y_true, y_pred, labels):
    cm = {true: {pred: 0 for pred in labels} for true in labels}

    for t, p in zip(y_true, y_pred):
        cm[t][p] += 1

    return cm

def pretty_print_cm(cm, labels):
    print("Confusion Matrix:")
    print("\t" + "\t".join(labels))
    
    for true in labels:
        row = [str(cm[true][pred]) for pred in labels]
        print(f"{true}\t" + "\t".join(row))

def compute_metrics(cm, labels):
    for label in labels:
        TP = cm[label][label]
        FP = sum(cm[other][label] for other in labels if other != label)
        FN = sum(cm[label][other] for other in labels if other != label)

        precision = TP / (TP + FP) if (TP + FP) else 0
        recall = TP / (TP + FN) if (TP + FN) else 0

        print(f"{label} -> Precision: {precision:.2f}, Recall: {recall:.2f}")

# Test data
labels = ['cat', 'dog', 'bird']

y_true = ['cat','dog','bird','cat','dog','bird','cat','dog','bird','cat','dog','bird','cat','dog','bird']
y_pred = ['cat','dog','cat','cat','bird','bird','dog','dog','bird','cat','dog','cat','bird','dog','bird']

cm = confusion_matrix(y_true, y_pred, labels)

pretty_print_cm(cm, labels)
print()
compute_metrics(cm, labels)
```
