```python
def tokenize(text):
    stop_words = {"a", "an", "the", "is", "in", "of", "and", "to"}
    punctuation = ".,!?;:"

    text = text.lower()
    for p in punctuation:
        text = text.replace(p, "")
    words = text.split()
    token = [w for w in words if w not in stop_words]
    return token 

text = "Jack and Jill went up the hill to fetch a pale of water, Jack fell down and broke his leg and Jill came tumbling after."
tokens = tokenize(text)
print("Tokens:", tokens)  


vocab = sorted(set(tokens))
print("Vocabulary:", vocab)

freq = {}
for token in tokens:
    freq[token] = freq.get(token, 0) + 1
print("Frequency:", freq)    
```
