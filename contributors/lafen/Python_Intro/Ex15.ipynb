```python
best_val_accuracy = 0.0
checkpoint_log = []

def update_checkpoint(epoch, val_acc, model_name):
    global best_val_accuracy

    if val_acc > best_val_accuracy:
        best_val_accuracy = val_acc

    checkpoint_log.append({
        'epoch': epoch,
        'val_accuracy': val_acc,
        'best_so_far': best_val_accuracy,
        'model': model_name
    })

# Simulated epochs
update_checkpoint(1, 0.70, "ModelX")
update_checkpoint(2, 0.75, "ModelX")
update_checkpoint(3, 0.73, "ModelX")
update_checkpoint(4, 0.80, "ModelX")
update_checkpoint(5, 0.78, "ModelX")

print(checkpoint_log)
```
