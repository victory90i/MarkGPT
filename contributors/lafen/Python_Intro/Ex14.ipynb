```python
def report_metrics(model_name, accuracy, loss, verbose=True, round_to=4):
    accuracy = round(accuracy, round_to)
    loss = round(loss, round_to)

    if verbose:
        print(f"Model: {model_name}")
        print(f"Accuracy: {accuracy}")
        print(f"Loss: {loss}")
    else:
        print(f"{model_name} | Accuracy: {accuracy}")

# Calls
report_metrics("ModelA", 0.923456, 0.23456)
print()

report_metrics("ModelB", 0.9123, 0.3)  # defaults
print()

report_metrics("ModelC", 0.87654, 0.45, verbose=False)
print()

report_metrics("ModelD", 0.987654, 0.12345, round_to=2)
```
