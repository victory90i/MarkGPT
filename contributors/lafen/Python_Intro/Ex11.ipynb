```python
def celsius_to_kelvin(numbers):
    kelvin = [(number + 273.15) for number in numbers]
    return kelvin

def normalisation(numbers):
    normalize = [((number - min(celsius_to_kelvin(numbers)))/ (max(celsius_to_kelvin(numbers)) - min(celsius_to_kelvin(numbers)))) for number in celsius_to_kelvin(numbers)]
    return normalize

temperature_reading = [10, 25, 37, 100, -5]
print(celsius_to_kelvin(temperature_reading))
print(normalisation(temperature_reading))
```
