```python
experiment_tracker = {}

def log_run(run_id, params, metrics):
    experiment_tracker[run_id] = {
        'params': params,
        'metrics': metrics
    }

def get_best_run(metric_name):
    best_run = None
    best_value = float('-inf')

    for run_id, data in experiment_tracker.items():
        value = data['metrics'].get(metric_name, 0)
        if value > best_value:
            best_value = value
            best_run = run_id

    return best_run

def summarise():
    for run_id, data in experiment_tracker.items():
        print(f"{run_id} | Params: {data['params']} | Metrics: {data['metrics']}")

# Log runs
log_run("run1", {'lr': 0.01}, {'accuracy': 0.85, 'loss': 0.3})
log_run("run2", {'lr': 0.001}, {'accuracy': 0.90, 'loss': 0.25})
log_run("run3", {'lr': 0.005}, {'accuracy': 0.88, 'loss': 0.28})
log_run("run4", {'lr': 0.0005}, {'accuracy': 0.92, 'loss': 0.22})

print("Best run:", get_best_run('accuracy'))
print()
summarise()
```
