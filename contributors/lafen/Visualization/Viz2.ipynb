```python
import numpy as np
import matplotlib.pyplot as plt
# One sensor reading at one time step
reading_a = np.array([38.1, 220.5, 1.50, 0.03]) # [temp, V, A, vibration]
reading_b = np.array([36.0, 221.0, 1.40, 0.02])

# Vector addition — element-wise
print("Sum:", reading_a + reading_b)

# Scalar multiplication — scale every element
print("Doubled:", reading_a * 2)

# The dot product of two vectors: sum of element-wise products
# In maths notation: a · b = a[0]*b[0] + a[1]*b[1] + ....
dot = np.dot(reading_a, reading_b)
print(f"Dot product: {dot:.4f}")

magnitude_a = np.linalg.norm(reading_a)
print(f"Magnitude of reading_a: {magnitude_a:.4f}")

# ─── COSINE SIMILARITY ────────────────────────────────────────────────
# Measures how similar two vectors are in direction (not magnitude)
# Value ranges from -1 (opposite) to +1 (identical direction)
cos_sim = np.dot(reading_a, reading_b) / (
np.linalg.norm(reading_a) * np.linalg.norm(reading_b))
print(f"Cosine similarity: {cos_sim:.4f}")
```
