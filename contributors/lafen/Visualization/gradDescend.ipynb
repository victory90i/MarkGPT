```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
     return x**2

def f_grad(x):
    return 2*x    
# Goal: find the x that minimizes f(x) = x^2 (answer: x = 0 )
# We start at x = 2.5 and walk downhill by following the negative gradient

learning_rate = 0.1 # how big a step to take each iteration
x = 2.5 # starting point
history = [] # track the path

x_vals = np.linspace(-3, 3, 300)
# x_vals
y = f(x_vals)

# Training 
for step in range(20):
    gradient = f_grad(x)
    x = x - learning_rate * gradient
    history.append((x, f(x)))
    print(f"step {step + 1:2d} x = {x :.4f} loss = {f(x):.6f}")
    
xs, ys = zip(*history)    
plt.figure(figsize = (10, 5))
plt.plot(x_vals, f(x_vals), color = "cyan", alpha = 0.7, linewidth = 2)
plt.scatter(xs, ys, c = range(len(xs)), cmap = "Reds", s = 60, zorder = 5)
for i, (xi, yi) in enumerate(history[:-1]):
    plt.annotate("", xy = history[i + 1], xytext = (xi, yi), arrowprops = dict(arrowstyle = "->", color = "blue", lw = 1.2))
plt.title("Gradient Descent: Walking Downhill to the Minimum ")
plt.xlabel("x (model parameter)")
plt.ylabel("f(x) = loss")
plt.savefig("gradient descent.png", dpi=150)
plt.show()
    
```
