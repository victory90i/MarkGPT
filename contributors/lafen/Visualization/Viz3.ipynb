```python
import numpy as np
import matplotlib.pyplot as plt
# from scipy import stats

np.random.seed(0)
true_temp = 37.0
sigma = 0.5
readings = np.random.normal(loc = true_temp, scale = sigma, size = 1000)

fig, ax = plt.subplots(figsize = (10, 5))

ax.hist(readings, bins = 40, density = True,  color = "#2563EB", edgecolor = "white", alpha = 0.7, label = "1000 simulated readeing")
# alpha means the opacity, how visible it is

# Overlay the theoritical PDF
```
