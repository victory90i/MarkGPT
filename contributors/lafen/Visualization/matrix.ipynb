```python
import numpy as np
# mat_A = np.array([
#                 [5,10,15],
#                 [2, 4, 6],
#                 [1, 2, 3]
# ])
# print(mat_A)

# Input matrix: shape (3, 4) — 3 samples, 4 features
X = np.array([
[38.1, 220.5, 1.50, 0.03], # sample 0
[36.0, 221.0, 1.40, 0.02], # sample 1
[40.3, 218.8, 1.68, 0.05], # sample 2
])
# Weight matrix: shape (4, 2) — maps 4 features to 2 outputs
np.random.seed(7)
W = np.random.randn(4, 2) * 0.1 # small random weights
# Matrix multiplication: (3×4) @ (4×2) = (3×2)
# for matrix multiplication of size (m1, m2) * (m3, m4), m2 must be = m3 and the solution will be a (m1, m4) matrix
output = X @ W # The @ operator is matrix multiply in Python
print("Input shape: ", X.shape) # (3, 4)
print("Weight shape:", W.shape) # (4, 2)
print("Output shape:", output.shape) # (3, 2)
print(output)
# This output is exactly what happens at one layer of a neural network.
# Each row of the output is the 'hidden representation' of one sample.

print("the transpose of W is:", W.T.shape)
```
