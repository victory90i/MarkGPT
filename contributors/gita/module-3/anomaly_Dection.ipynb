```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


np.random.seed(42)
NORMAL  = '#378ADD'
ANOMALY = '#E24B4A'
BOUND   = '#BA7517'
BG      = '#F8F8F6'
```

```python
# MRHOD-1 Z-SCORE (univariate, works per feature in multivariate case)
# z=(x-u)/sigma

# Flag as anomaly if |z| > threshold (commonly 3.0)
# Assumption os that .. data is approximately Gaussian
```

```python
def zscore_anomaly_detector(data:np.ndarray, threshold:float=3.0):
    if data.ndim==1:
        mu=data.mean()
        sigma=data.std(ddof=1)
        z=np.abs((data-mu)/(sigma + 1e-9))
    else:
        mu=data.mean(axis=0)
        sigma=data.std(axis=0, ddof=1)
        z=np.abs((data-mu)/(sigma+1e-9))
        z=z.max(axis=1)  # worst z score accross features
        return z, z>threshold
```

```python
def iqr_anomaly_detector(data:np.ndarray, k:float=1.5):
    # k is thhe multiplier for the iqr,
    # lower, uppper => fence values
    # is_anomaly : boolean array
    Q1=np.percentile(data, 25)
    Q3=np.percentile(data, 75)
    IQR=Q3-Q1
    lower=Q1-k*IQR
    upper=Q3+k*IQR
    return lower, upper, (data < lower) | (data > upper)
```

```python
def main_result():
    x_normal=np.random.multivariate_normal([0,0],[[1, 0.6], [0.6, 1]], 200)
    x_anom   = np.array([[5, 5], [-4, 4], [4, -4], [-5, -5],
                          [3.5, -3], [0, 6]])
    # print(f"the normal vlaues distributed are: {x_normal}")
    x_all=np.vstack([x_normal, x_anom])
    y_true=np.array([0]*200 + [1]*6)
    z_score, z_anom=zscore_anomaly_detector(x_all, threshold=3.0)
    print("\n==Z-score ===")
    print(f" Detected {z_anom.sum()} anomalies out of  {len(z_anom)} points")
    plt.figure(figsize=(14, 10))
# SUBPLOT 1 : NORMAL DATA + ANOMALIES
    plt.subplot(2,2,1)

    sns.scatterplot(
    x=x_normal[:,0],
    y=x_normal[:, 1],
    label="Normal Data",
    color="red"

)
    sns.scatterplot(
        x=x_anom[:,0],
        y=x_anom[:,1],
        label="Injected Anomalies",
        s=120,
        color="green"
    )
    
    
    plt.axhline(0, linestyle="--")
    plt.axvline(0, linestyle="--")

    plt.title("Multivariate Gaussian Distribution")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    # SUBPLOT 2: ALL DATA WIT DETECTED ANOMALIES 
    plt.subplot(2, 2, 2)
    sns.scatterplot(
        x=x_all[:, 0],
        y=x_all[:, 1],
        hue=z_anom,
        s=70
    )
    plt.title("Detected Anomalies using z-score")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")

    #  --IQR (1-D DEMO ONN FEATURE 0 ONLY)
    lower, upper, iqr_anom=iqr_anomaly_detector(x_all[:,0])
    print("IQR ON FEATURE X ONLY")
    print(f"Fences: [{lower:.3f}, {upper:.3f}]")
    print(f"Detected {iqr_anom.sum()} anomalies on x, ")
```
