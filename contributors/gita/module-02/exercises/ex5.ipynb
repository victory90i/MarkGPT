# Q5. SciPy Normality Check

```python
import numpy as np
import scipy.stats as stats

np.random.seed(0)

datasets={
    'Normal(0,1)': np.random.normal(0, 1, 500),
    'Exponential(1)':np.random.exponential(1, 500),
    'Uniform(0,1)':np.random.uniform(0,1,500),
}

print(f"{'Dataset':<20} {'skewness':>10}  {'p-value':>12}  {'Verdict':>12}")
print('-' *60)
for name, data in datasets.items():
    skew=stats.skew(data)
    _, p=stats.normaltest(data)
    verdict='NORMAL' if p>0.05 else 'NOT NORMAL'
    print(f'{name:<20} {skew:>10.4f} {p:>12.6f} {verdict:>12}')
```
