## Q2. NumPy Arithmetic Playground

```python
import numpy as np

x=np.linspace(0, 4*np.pi,20)
x
```

```python
y=np.sin(x)
y
```

```python
# (a) np.argmax returns the INDEX of the largest value, not the value itself

max_idx=np.argmax(y)
```

```python
max_idx
```

```python
 # (b) np.mean and np.std operate on the whole array in one call
print(f"Mean = {np.mean(y):.4f}, Std={np.std(y):.4f}")
```

```python
# (c) Boolean mask: y > 0.5 creates a True/False array of the same shape as y
# Passing that mask back into y[] selects only the True positions
```

```python
count_above=np.sum(y>0.5) # True counts as 1 , False as 0
print(f"Values exceeding 0.5: {count_above}")
```
