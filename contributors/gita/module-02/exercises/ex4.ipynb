```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# “Give me 200 fake exam scores where the average is 70, and most scores fall within a reasonable spread of 12.”
np.random.seed(42)  # seed for reproducibility

# KDE (Kernel Density Estimation) is a method for estimating the probability distribution of a dataset by creating a 
# smooth curve instead of using discrete bars like a histogram. It works by placing a small smooth function (kernel) on each 
# data point and summing them to form a continuous density curve. This makes it easier to see patterns such as where values cluster
# or how data is spread. In practice, KDE helps 
# 1visualize the overall shape of data without being affected by arbitrary bin choices in histograms.


df=pd.DataFrame({'exams_score':np.random.normal(70, 2, 200)})
df.sample(100)

# plt.subplots creates a Figure and an array of Axes objects
# figsize=(12,4) sets width=12 inches, height=4 inches

fig, axes=plt.subplots(1, 2, figsize=(12, 4))

# pass ax= to tell seaborn which subplot to draw on
sns.histplot(data=df, x='exams_score', color='steelblue', ax=axes[0])
axes[0].set_title("Distribution with KDE Curve")
axes[0].set_xlabel('Exam Score')

sns.boxplot(data=df, y='exams_score', color='gold', ax=axes[1])

axes[1].set_title('Box Plot of Scores')
plt.tight_layout() # prevents subplots from overlapping
plt.savefig('toolkit_exercise.png', dpi=120, bbox_inches='tight')
plt.show()
```
