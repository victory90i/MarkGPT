```python
import numpy as np
import matplotlib.pyplot as plt

# DERIVATIVE INTUITION

# f(x)=x**2 => f'(x)=2x
# at any x the slope of the curve is 2x
def f(x):
    return x**2

f(2)
```

```python
def f_grad(x):
    return 2*x
x_vals=np.linspace(-3,3,300)
x_vals

plt.figure(figsize=(10,5))
plt.plot(x_vals, f(x_vals), color="#2563EB", linewidth=2,label="loss: f(x)=x^2")
# Mark gradient at x=2
x0=2.0;
grad=f_grad(x0)


tangent=grad * (x_vals-x0) + f(x0)

plt.plot(x_vals, tangent, color="#E74C3C", linestyle="--", alpha=0.7, label=f"Tangent at x={x0} (slope={grad})")

plt.scatter([x0], [f(x0)], s=100, color="red", zorder=5)
plt.ylim(-0.5, 10)
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.title("Derivative = Slope of Tangent = Direction to Walk")
plt.legend()
    
```

```python
learning_rate=0.1;
x=2.5;
history=[]
for step in range(20):
    # print(step)
    gradient=f_grad(x)
    x=x-learning_rate*gradient
    history.append((x,f(x)))
    # print(history)
    # print(x)
# Visualize the descent path
xs, ys=zip(*history)
plt.figure(figsize=(10,5))
plt.plot(x_vals, f(x_vals), color="red", linewidth=2)
plt.scatter(xs, ys, c=range(len(xs)), cmap="Reds", s=60, zorder=5)
for i, (xi, yi) in enumerate(history[:-1]):
    plt.annotate("", xy=history[i+1], xytext=(xi, yi), arrowprops=dict(arrowstyle="->", color="green", lw=1.2))
    plt.title("Gradient Descent: Walking Downhill to the Minimum")
plt.xlabel("x (model parameter)")
plt.ylabel("f(x) = loss")
plt.savefig("gradient_descent.png", dpi=150)
plt.show()
```

```python
np.random.seed(1)
X_train=np.linspace(0, 10,50)
y_train=2.5*X_train +30 + np.random.normal(0, 2, 50)

m=0.0
b=0.0
lr=0.001
n=len(X_train)

losses=[]

for epoch in range(1000000):
    # Forward pass: make predictions
    y_pred=m*X_train+b;
    # compute loss (Mean squared Error)
    loss=np.mean((y_pred-y_train)**2)
    print(f"Loss: {loss}")
    losses.append(loss)
    # Compute gradients (partial derivatives of MSE w.r.t. m and b)
    d_m = (2/n) * np.sum((y_pred - y_train) * X_train)
    d_b=(2/n)*np.sum(y_pred-y_train)
    # update parameters - step opposite to the gradiwnt
    m=m-lr*d_m
    b=b-lr*d_b
    if epoch%2000==0:
        print(f"Epoch {epoch:4d}  Loss={loss:.3f} m={m:.3f} b={b:.3f}")
print(f"\nFinal: m={m:.3f} b={b:.3f}")
print(f"True values: m=2.5 b=30.0")


    
```
