## Q3. Pandas DataFrame Surgery

```python
import numpy as np
import pandas as pd



df=pd.DataFrame({
    'name':['Alice', 'Bob', 'Carlson', 'Diana','Eve'],
    'score':[88, 72, 95, 61, 83],
    'passed':[True, True, True, False, True]
})

df
```

```python
# np.select is the vectorised if/elif/else — much faster than a Python loop
# conditions are checked in ORDER; the first True match wins

conditions=[df['score']>=90, df['score']>=80]
conditions
choices=['A', 'B']
df['grades']=np.select(conditions, choices, default='C')
df
```

```python
passing=df[df['passed']]
passing
```
