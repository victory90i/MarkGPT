```python
import numpy as np
import pandas as pd
np.random.seed(7)
df = pd.DataFrame({
 'height_cm': np.random.normal(170, 10, 100),
 'weight_kg': np.random.normal(70, 15, 100),
 'age': np.random.randint(18, 65, 100),
 # np.random.choice samples randomly from a given list
 'category': np.random.choice(['A', 'B', 'C'], 100),
})

# Default describe() — only numeric columns
print(df.describe())
# include='all' adds categorical statistics alongside numeric ones
print(df.describe(include='all'))
# Extra columns for 'category': unique=3, top='B' (or whichever is most common),
# freq=number of times top appears. Numeric columns get NaN for unique/top/freq
```
