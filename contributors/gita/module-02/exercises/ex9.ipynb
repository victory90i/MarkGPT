```python
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D', 'F']
sizes = [8, 14, 12, 4, 2]
explode = [0, 0, 0, 0, 0.1] # pull the F slice outward by 0.1
# Colorblind-safe palette: uses blues, oranges, purples — avoids red/green
colors = ['#2166AC', '#74ADD1', '#ABD9E9', '#FEE090', '#F46D43']
fig, ax = plt.subplots(figsize=(7, 6))
ax.pie(
 sizes,
 labels=labels,
 autopct='%1.1f%%', # show percentage with 1 decimal
 explode=explode,
 colors=colors,
 startangle=90, # 90 = first slice starts at 12 o'clock
 wedgeprops={'edgecolor': 'white', 'linewidth': 1.5}
)
ax.set_title('Grade Distribution — Class of 40 Students', fontweight='bold')
ax.legend(title='Grade', loc='lower right')
plt.tight_layout()
plt.show()
```
