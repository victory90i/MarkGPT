```python
import numpy as np, pandas as pd, scipy.stats as stats
import matplotlib.pyplot as plt, seaborn as sns

# (a) load CSV of housing prices -> pandas
df=pd.read_csv('housing.csv')
# (b) 95th percentile -> numpy
data=np.random.normal(0,1,1000)
print(np.percentile(data, 95))

# (c) Correlation heatmap-> seaborn
sns.heatmap(df.corr((), annot=True, cmap='coolwarm'))
# (d) 1000 Poison random numbers-> numpy
poisson_samples=np.random.poisson(lam=5, size=1000)

#(e) t-test->scipy
a, b=np.random.normal(0,1,50), np.random.normal(0.3, 1, 50)
t, p=stats.ttest_ind(a, b, equal_var=False)
print(f't={t:.3f}, p={p:.4f}')
# (f) Scatter with precise ticks -> Matplotlib
fig, ax = plt.subplots()
ax.scatter(data[:50], data[50:100], alpha=0.5)
ax.set_xticks(np.arange(-3, 4, 1))
plt.show()

```
